```python
def summarize_source_code_issues_in_all_pair_res(
        all_pair_res: dict) -> pd.DataFrame:
    rows = []

    for pair_key, res_xreg in tqdm(
            all_pair_res.items(), desc="Checking SOURCE_CODE in all_pair_res"):
        # if res_xreg isn't a dict for some reason
        if not isinstance(res_xreg, dict):
            rows.append({
                "pair": pair_key,
                "df_key": None,
                "status": "res_xreg_not_dict",
                "n_rows": None,
                "unique_codes": None,
                "n_missing": None,
            })
            continue

        for df_key, obj in res_xreg.items():
            if not isinstance(obj, pd.DataFrame):
                continue

            df = obj
            n_rows = len(df)

            if "SOURCE_CODE" not in df.columns:
                rows.append({
                    "pair": pair_key,
                    "df_key": df_key,
                    "status": "MISSING_COLUMN",
                    "n_rows": n_rows,
                    "unique_codes": None,
                    "n_missing": None,
                })
                continue

            n_missing = int(df["SOURCE_CODE"].isna().sum())
            unique_codes = sorted(
                map(str, df["SOURCE_CODE"].dropna().unique().tolist()))

            status = "OK"
            if n_missing > 0:
                status = "HAS_NA"

            rows.append({
                "pair": pair_key,
                "df_key": df_key,
                "status": status,
                "n_rows": n_rows,
                "unique_codes": unique_codes,
                "n_missing": n_missing,
            })

    out = pd.DataFrame(rows)

    # nice ordering
    if not out.empty:
        out = out.sort_values(["pair", "df_key",
                               "status"]).reset_index(drop=True)

    return out


df_sc_check = summarize_source_code_issues_in_all_pair_res(all_pair_res)
df_sc_check
```

### Fixed glacier hold-out split (spatial generalization):
#### Calculate directional “shift as seen from the source model

```python
from sklearn.preprocessing import StandardScaler
from sklearn.covariance import LedoitWolf

MONTHLY_COLS = [
    "t2m",
    "tp",
    "slhf",
    "sshf",
    "ssrd",
    "fal",
    "str",
    "ELEVATION_DIFFERENCE",
]
STATIC_COLS = ["aspect", "slope", "svf"]
FEATURE_COLS = MONTHLY_COLS + STATIC_COLS


def _to_feature_matrix(
    df: pd.DataFrame,
    feature_cols: list[str],
    dropna: bool = True,
) -> np.ndarray:
    """Return X (n_samples, n_features) as float64."""
    missing = [c for c in feature_cols if c not in df.columns]
    if missing:
        raise KeyError(f"Missing feature columns in df: {missing}")

    X = df[feature_cols].to_numpy(dtype=np.float64, copy=True)

    if dropna:
        mask = np.isfinite(X).all(axis=1)
        X = X[mask]

    return X


def mahalanobis_mean_distance(
    X_ref: np.ndarray,
    X_other: np.ndarray,
    shrinkage: bool = True,
    eps: float = 1e-12,
) -> float:
    """
    Mahalanobis distance between means, using covariance estimated on X_ref
    (optionally LedoitWolf shrinkage for stability).
    """
    if X_ref.size == 0 or X_other.size == 0:
        return np.nan

    mu_ref = X_ref.mean(axis=0)
    mu_oth = X_other.mean(axis=0)
    dmu = (mu_ref - mu_oth).reshape(-1, 1)

    if shrinkage:
        cov = LedoitWolf().fit(X_ref).covariance_
    else:
        cov = np.cov(X_ref, rowvar=False)

    # Regularize just in case
    cov = cov + np.eye(cov.shape[0]) * eps

    try:
        inv_cov = np.linalg.inv(cov)
    except np.linalg.LinAlgError:
        inv_cov = np.linalg.pinv(cov)

    d2 = float(dmu.T @ inv_cov @ dmu)
    return float(np.sqrt(max(d2, 0.0)))


def energy_distance(
    X: np.ndarray,
    Y: np.ndarray,
    max_samples: int = 5000,
    seed: int = 0,
) -> float:
    """
    Energy distance between two multivariate samples.

    Uses a subsampling cap for speed. Complexity is O(n*m + n^2 + m^2) so keep caps moderate.
    Returns sqrt(ed2) where:
        ed2 = 2 E||X-Y|| - E||X-X'|| - E||Y-Y'||
    """
    if X.size == 0 or Y.size == 0:
        return np.nan

    rng = np.random.default_rng(seed)

    def _subsample(A: np.ndarray) -> np.ndarray:
        if A.shape[0] <= max_samples:
            return A
        idx = rng.choice(A.shape[0], size=max_samples, replace=False)
        return A[idx]

    Xs = _subsample(X)
    Ys = _subsample(Y)

    # Pairwise Euclidean distances without materializing (n,m,feat)
    # Use (x - y)^2 = x^2 + y^2 - 2 x·y trick.
    def _mean_pairwise_dist(A: np.ndarray, B: np.ndarray) -> float:
        a2 = np.sum(A * A, axis=1, keepdims=True)  # (n,1)
        b2 = np.sum(B * B, axis=1, keepdims=True).T  # (1,m)
        # squared distances
        d2 = a2 + b2 - 2.0 * (A @ B.T)
        d2 = np.maximum(d2, 0.0)
        d = np.sqrt(d2)
        return float(d.mean())

    # For within-sample, exclude diagonal? Energy distance definition uses independent copies,
    # so including diagonal has negligible effect for large n; we can include for simplicity.
    EXY = _mean_pairwise_dist(Xs, Ys)
    EXX = _mean_pairwise_dist(Xs, Xs)
    EYY = _mean_pairwise_dist(Ys, Ys)

    ed2 = 2.0 * EXY - EXX - EYY
    return float(np.sqrt(max(ed2, 0.0)))
```

```python
def compute_topoclimatic_distances(
    df_src: pd.DataFrame,
    df_pool: pd.DataFrame,
    df_holdout: pd.DataFrame,
    feature_cols: list[str],
    seed: int,
    energy_max_samples: int = 4000,
) -> dict:
    """
    Compute standardized topoclimatic distances between distributions:
      - src vs pool
      - src vs holdout
      - pool vs holdout
      - src vs target_all (pool ∪ holdout)
    """
    # Combine target
    df_tgt_all = pd.concat([df_pool, df_holdout], axis=0, ignore_index=True)

    Xs = _to_feature_matrix(df_src, feature_cols, dropna=True)
    Xp = _to_feature_matrix(df_pool, feature_cols, dropna=True)
    Xh = _to_feature_matrix(df_holdout, feature_cols, dropna=True)
    Xt = _to_feature_matrix(df_tgt_all, feature_cols, dropna=True)

    if Xs.size == 0 or Xp.size == 0 or Xh.size == 0 or Xt.size == 0:
        return {
            "n_src": int(Xs.shape[0]),
            "n_pool": int(Xp.shape[0]),
            "n_holdout": int(Xh.shape[0]),
            "n_tgt_all": int(Xt.shape[0]),
            "D_maha_src_pool": np.nan,
            "D_maha_src_holdout": np.nan,
            "D_maha_pool_holdout": np.nan,
            "D_maha_src_tgt_all": np.nan,
            "D_energy_src_pool": np.nan,
            "D_energy_src_holdout": np.nan,
            "D_energy_pool_holdout": np.nan,
            "D_energy_src_tgt_all": np.nan,
        }

    scaler = StandardScaler()
    Xs_z = scaler.fit_transform(Xs)
    Xp_z = scaler.transform(Xp)
    Xh_z = scaler.transform(Xh)
    Xt_z = scaler.transform(Xt)

    # Mahalanobis between means (cov from src unless otherwise noted)
    d_maha_src_pool = mahalanobis_mean_distance(Xs_z, Xp_z, shrinkage=True)
    d_maha_src_hold = mahalanobis_mean_distance(Xs_z, Xh_z, shrinkage=True)
    d_maha_pool_hold = mahalanobis_mean_distance(Xp_z, Xh_z, shrinkage=True)
    d_maha_src_tgt = mahalanobis_mean_distance(Xs_z, Xt_z, shrinkage=True)

    # Energy distance (distribution-level)
    d_en_src_pool = energy_distance(Xs_z,
                                    Xp_z,
                                    max_samples=energy_max_samples,
                                    seed=seed + 1)
    d_en_src_hold = energy_distance(Xs_z,
                                    Xh_z,
                                    max_samples=energy_max_samples,
                                    seed=seed + 2)
    d_en_pool_hold = energy_distance(Xp_z,
                                     Xh_z,
                                     max_samples=energy_max_samples,
                                     seed=seed + 3)
    d_en_src_tgt = energy_distance(Xs_z,
                                   Xt_z,
                                   max_samples=energy_max_samples,
                                   seed=seed + 4)

    return {
        "n_src": int(Xs.shape[0]),
        "n_pool": int(Xp.shape[0]),
        "n_holdout": int(Xh.shape[0]),
        "n_tgt_all": int(Xt.shape[0]),
        "D_maha_src_pool": float(d_maha_src_pool),
        "D_maha_src_holdout": float(d_maha_src_hold),
        "D_maha_pool_holdout": float(d_maha_pool_hold),
        "D_maha_src_tgt_all": float(d_maha_src_tgt),
        "D_energy_src_pool": float(d_en_src_pool),
        "D_energy_src_holdout": float(d_en_src_hold),
        "D_energy_pool_holdout": float(d_en_pool_hold),
        "D_energy_src_tgt_all": float(d_en_src_tgt),
    }
```

```python
MONTHLY_COLS = [
    "t2m",
    "tp",
    "slhf",
    "sshf",
    "ssrd",
    "fal",
    "str",
    "ELEVATION_DIFFERENCE",
]
STATIC_COLS = ["aspect", "slope", "svf"]
FEATURE_COLS = MONTHLY_COLS + STATIC_COLS

pair_splits = {}

pbar = tqdm(list(all_pair_res.items()), desc="Pool/holdout + distances")
for key, res in pbar:
    src, tgt = key.split("_to_")
    pbar.set_postfix_str(key)

    # source and target frames
    df_src = res["df_train"].copy()

    df_tgt = res["df_test"].copy()
    df_tgt = df_tgt[df_tgt["SOURCE_CODE"] == tgt].copy()  # safety

    if len(df_src) == 0 or len(df_tgt) == 0:
        pair_splits[key] = {
            "source": src,
            "target": tgt,
            "error": "empty src or tgt"
        }
        continue

    # cluster-stratified glacier split on target
    df_pool, df_holdout, holdout_glaciers, pool_glaciers, gfeat, split_summary = (
        holdout_split_cluster_stratified(
            df_tgt,
            holdout_frac=0.30,
            seed=cfg.seed,
            n_clusters=6,
        ))

    # compute distances (includes src -> (pool+holdout) = tgt_all)
    dist = compute_topoclimatic_distances(
        df_src=df_src,
        df_pool=df_pool,
        df_holdout=df_holdout,
        feature_cols=FEATURE_COLS,
        seed=cfg.seed,
        energy_max_samples=4000,
    )

    pair_splits[key] = {
        "source": src,
        "target": tgt,
        "df_src": df_src,
        "df_target": df_tgt,
        "df_pool": df_pool,
        "df_holdout": df_holdout,
        "pool_glaciers": pool_glaciers,
        "holdout_glaciers": holdout_glaciers,
        "gfeat": gfeat,
        "split_summary": split_summary,
        "distances": dist,
    }

    print(f"{key}: "
          f"D_energy(src, tgt_all)={dist['D_energy_src_tgt_all']:.3f} | "
          f"D_maha(src, tgt_all)={dist['D_maha_src_tgt_all']:.3f} | "
          f"D_energy(src, hold)={dist['D_energy_src_holdout']:.3f} | "
          f"D_energy(pool, hold)={dist['D_energy_pool_holdout']:.3f} | ")
```

### Build static assets once (source + target holdout)

```python
import hashlib


def split_signature(glaciers):
    g = sorted(map(str, glaciers))
    return hashlib.md5((",".join(g)).encode("utf-8")).hexdigest()[:10]


for key, d in pair_splits.items():
    if "error" in d:
        continue
    print(key, "holdout_sig:", split_signature(d["holdout_glaciers"]),
          "n_hold:", len(d["holdout_glaciers"]))
```

```python
static_assets_by_pair = {}
errors_static_assets = {}

pbar = tqdm(list(pair_splits.items()), desc="Building static TL assets")
for key, d in pbar:
    if "error" in d:
        errors_static_assets[key] = d["error"]
        continue

    src = d["source"]
    tgt = d["target"]
    pbar.set_postfix_str(f"{src}->{tgt}")

    res_xreg = all_pair_res[key]
    holdout_glaciers = d["holdout_glaciers"]

    sig = split_signature(holdout_glaciers)
    cache_dir_pair = CACHE_DIR / key / f"holdout_{sig}"
    cache_dir_pair.mkdir(parents=True, exist_ok=True)

    def _build(force_recompute: bool):
        return build_static_tl_assets_source_and_holdout(
            cfg=cfg,
            res_xreg=res_xreg,
            target_code=tgt,
            source_code=src,
            holdout_glaciers=holdout_glaciers,
            MONTHLY_COLS=MONTHLY_COLS,
            STATIC_COLS=STATIC_COLS,
            cache_dir=cache_dir_pair,
            force_recompute=force_recompute,
            show_progress=True,
        )

    try:
        # first try: load if available
        static_assets = _build(force_recompute=False)
        static_assets_by_pair[key] = static_assets

    except Exception as e:
        msg = repr(e)

        # Only auto-recompute for the specific cache/lookup failure pattern
        should_retry = ("Missing SOURCE_CODE for ds key" in msg) or isinstance(
            e, KeyError)

        if should_retry:
            try:
                # rebuild cache cleanly
                static_assets = _build(force_recompute=True)
                static_assets_by_pair[key] = static_assets
            except Exception as e2:
                errors_static_assets[
                    key] = f"initial={msg} | recompute_failed={repr(e2)}"
        else:
            errors_static_assets[key] = msg

print("\nDone.")
print("Built assets for:", list(static_assets_by_pair.keys()))
if errors_static_assets:
    print("\nErrors:")
    for k, v in errors_static_assets.items():
        print(" ", k, ":", v)
```

```python
pool_capacity_by_pair = {}
errors_pool_capacity = {}

pbar = tqdm(list(pair_splits.items()), desc="Computing pool capacity (G/Y/M)")
for key, d in pbar:
    if "error" in d:
        errors_pool_capacity[key] = d["error"]
        continue

    src = d["source"]
    tgt = d["target"]
    pbar.set_postfix_str(f"{src}->{tgt}")

    pool_glaciers = d["pool_glaciers"]

    # monthly target df (already filtered to target in earlier steps)
    df_tgt_monthly = d["df_target"]
    df_pool_monthly = df_tgt_monthly[df_tgt_monthly["GLACIER"].isin(
        pool_glaciers)].copy()

    if len(df_pool_monthly) == 0:
        errors_pool_capacity[key] = "empty df_pool_monthly"
        continue

    G_max = int(df_pool_monthly["GLACIER"].nunique())
    Y_max = int(df_pool_monthly.groupby("GLACIER")["YEAR"].nunique().max())

    # raw stake df for the target RGI region (needed for rows_per_gy)
    # Here we use TARGET_REGION_META[target]["rgi_region_id"] to get "06"/"08"/"11"
    rid = TARGET_REGION_META[tgt]["rgi_region_id"]  # "06", "08", "11"
    df_tgt_raw = dfs[rid].copy()
    df_tgt_raw = df_tgt_raw[df_tgt_raw["GLACIER"].isin(pool_glaciers)].copy()

    if len(df_tgt_raw) == 0:
        errors_pool_capacity[
            key] = f"empty raw df for rid={rid} after filtering pool_glaciers"
        continue

    rows_per_gy = df_tgt_raw.groupby(["GLACIER", "YEAR"]).size()
    M_p25 = int(rows_per_gy.quantile(0.25))
    M_p50 = int(rows_per_gy.median())
    M_p90 = int(rows_per_gy.quantile(0.90))
    M_max = int(rows_per_gy.max())
    M_min = int(rows_per_gy.min())

    pool_capacity_by_pair[key] = {
        "source": src,
        "target": tgt,
        "rid": rid,
        "G_max": G_max,
        "Y_max": Y_max,
        "M_min": M_min,
        "M_p25": M_p25,
        "M_p50": M_p50,
        "M_p90": M_p90,
        "M_max": M_max,
    }

    print("\n" + "=" * 80)
    print(f"POOL CAPACITY | {key}")
    print("G_max:", G_max)
    print("Y_max (max years on a glacier):", Y_max)
    print("Rows per glacier-year (raw format): min", M_min, "| p25", M_p25,
          "| median", M_p50, "| p90", M_p90, "| max", M_max)

# optional: tabular summary
df_pool_capacity = pd.DataFrame.from_dict(
    pool_capacity_by_pair,
    orient="index").reset_index().rename(columns={"index": "pair"})
display(df_pool_capacity)

if errors_pool_capacity:
    print("\nErrors:")
    for k, v in errors_pool_capacity.items():
        print(" ", k, ":", v)
```

```python
def compute_topoclimatic_distance_src_to_ft(
    df_src: pd.DataFrame,
    df_ft: pd.DataFrame,
    feature_cols: list[str],
    seed: int,
    energy_max_samples: int = 4000,
) -> dict:
    """
    Distances for one monitoring subset:
      - src vs ft (Mahalanobis-between-means + Energy distance)
    Standardization: fit StandardScaler on src, transform ft.
    """
    Xs = _to_feature_matrix(df_src, feature_cols, dropna=True)
    Xf = _to_feature_matrix(df_ft, feature_cols, dropna=True)

    if Xs.size == 0 or Xf.size == 0:
        return {
            "n_src": int(Xs.shape[0]),
            "n_ft": int(Xf.shape[0]),
            "D_maha_src_ft": np.nan,
            "D_energy_src_ft": np.nan,
        }

    scaler = StandardScaler()
    Xs_z = scaler.fit_transform(Xs)
    Xf_z = scaler.transform(Xf)

    d_maha = mahalanobis_mean_distance(Xs_z, Xf_z, shrinkage=True)
    d_en = energy_distance(Xs_z, Xf_z, max_samples=energy_max_samples, seed=seed + 101)

    return {
        "n_src": int(Xs.shape[0]),
        "n_ft": int(Xf.shape[0]),
        "D_maha_src_ft": float(d_maha),
        "D_energy_src_ft": float(d_en),
    }
```

```python
# -----------------------
# Budgets & seeds (global)
# -----------------------
G_set = [1, 2, 3, 5, 10, 15, 19]  # (19 is max accross all regions)
Y_set = [1, 2, 3, 5, 10, 20, 30, 35]
BUDGETS = [{"G": int(g), "Y": int(y)} for g in G_set for y in Y_set]

N_SEEDS = 10
SEEDS = list(range(1, N_SEEDS + 1))

print("Total budget points:", len(BUDGETS))
print(pd.DataFrame(BUDGETS).sort_values(["G", "Y"]).head(12))
print("Seeds:", SEEDS)

# -----------------------------------
# Build finetune assets for ALL pairs
# -----------------------------------
assets_all_by_pair = {}  # key -> dict of assets for that pair
errors_assets_by_pair = {}  # key -> list of errors (optional)
skipped_tasks_by_pair = {}  # key -> list of skipped (G,Y,seed) (optional)

pairs_iter = tqdm(list(pair_splits.items()),
                  desc="Pairs (build finetune assets)")
for pair_key, d in pairs_iter:
    if "error" in d:
        continue

    src = d["source"]
    tgt = d["target"]
    pairs_iter.set_postfix_str(f"{src}->{tgt}")

    # Require static assets already computed
    if pair_key not in static_assets_by_pair:
        errors_assets_by_pair.setdefault(
            pair_key, []).append("missing static_assets_by_pair")
        continue

    # Pair-specific inputs
    res_xreg = all_pair_res[pair_key]
    static_assets = static_assets_by_pair[pair_key]

    # pair+split-specific cache dir (prevents "why does it recompute" issues)
    sig = split_signature(d["holdout_glaciers"])
    cache_dir_pair = CACHE_DIR / pair_key / f"holdout_{sig}"
    cache_dir_pair.mkdir(parents=True, exist_ok=True)

    # Pool rows (monthly) for sampling monitoring subsets
    pool_glaciers = d["pool_glaciers"]
    df_pool_rows = d["df_target"][d["df_target"]["GLACIER"].isin(
        pool_glaciers)].copy()

    if len(df_pool_rows) == 0:
        errors_assets_by_pair.setdefault(pair_key,
                                         []).append("empty df_pool_rows")
        continue

    # Capacity (used only to decide skipping)
    G_max = int(df_pool_rows["GLACIER"].nunique())
    Y_max = int(df_pool_rows.groupby("GLACIER")["YEAR"].nunique().max())

    # Build tasks for this pair
    TASKS = [(b, seed) for b in BUDGETS for seed in SEEDS]

    assets_all = {}
    skipped = []
    errs = []

    task_iter = tqdm(TASKS, desc=f"Assets {pair_key}", leave=False)
    for b, seed in task_iter:
        G = int(b["G"])
        Y = int(b["Y"])

        # skip budgets that can't be satisfied for this target pool
        # (avoids sample_monitoring_subset_from_pool crashing)
        if (G > G_max) or (Y > Y_max):
            skipped.append({
                "G":
                G,
                "Y":
                Y,
                "seed":
                seed,
                "reason":
                f"exceeds pool capacity (G_max={G_max}, Y_max={Y_max})"
            })
            continue

        try:
            df_ft, chosen, _ = sample_monitoring_subset_from_pool(
                df_pool=df_pool_rows,
                G=G,
                Y=Y,
                seed=seed,
                glacier_pick_method="random",
                year_pick_method="earliest_block",
                enforce_full_Y_if_possible=True,
            )

            exp_key = f"TL_{src}_to_{tgt}_G{G}_Y{Y}_seed{seed}"
            
            dist_ft = compute_topoclimatic_distance_src_to_ft(
                    df_src=res_xreg["df_train"],
                    df_ft=df_ft,
                    feature_cols=FEATURE_COLS,
                    seed=seed,
                    energy_max_samples=4000,
                )

            assets_one = build_budget_assets_finetune_only(
                cfg=cfg,
                res_xreg=res_xreg,
                static_assets=static_assets,
                df_ft=df_ft,
                exp_key=exp_key,
                MONTHLY_COLS=MONTHLY_COLS,
                STATIC_COLS=STATIC_COLS,
                cache_dir=cache_dir_pair,  # <-- pair-specific cache
                force_recompute=False,
                val_ratio=0.2,
                show_progress=False,
                seed=seed,
            )
            assets_one[exp_key]["distances_ft"] = dist_ft

            assets_all.update(assets_one)

        except Exception as e:
            errs.append({"G": G, "Y": Y, "seed": seed, "error": repr(e)})

    assets_all_by_pair[pair_key] = assets_all
    if skipped:
        skipped_tasks_by_pair[pair_key] = skipped
    if errs:
        errors_assets_by_pair[pair_key] = errs

    print(f"\n{pair_key}: built={len(assets_all)} | "
          f"skipped={len(skipped)} | errors={len(errs)} | "
          f"G_max={G_max}, Y_max={Y_max}")

print("\nDone.")
print("Pairs with assets:", list(assets_all_by_pair.keys()))
print("Total assets (all pairs):",
      sum(len(v) for v in assets_all_by_pair.values()))
```

### Compute E_ZERO:
E_ZERO = error of the CH baseline model evaluated on the fixed ISL holdout dataset (unseen glaciers), with no finetuning

```python
E0_rows = []
preds_zero_by_pair = {}  # optional: store preds dfs

MAKE_PLOTS = False  # set True if you want a figure per pair

for src, tgt in tqdm(PAIRS, desc="E_ZERO for all directed pairs"):
    pair_key = f"{src}_to_{tgt}"

    if pair_key not in static_assets_by_pair:
        E0_rows.append({
            "pair": pair_key,
            "source": src,
            "target": tgt,
            "error": "missing static_assets"
        })
        continue
    if src not in baseline_models:
        E0_rows.append({
            "pair": pair_key,
            "source": src,
            "target": tgt,
            "error": "missing baseline_model"
        })
        continue

    static_assets = static_assets_by_pair[pair_key]
    model_src = baseline_models[src]

    # optional plot axis
    ax = None
    fig = None
    if MAKE_PLOTS:
        fig, ax = plt.subplots(1, 1, figsize=(6, 6))

    try:
        metrics_zero, df_preds_zero, _, _ = evaluate_one_model_TL(
            cfg=cfg,
            model=model_src,
            device=device,
            tl_assets_for_key=static_assets,
            ax=ax,
            title=f"E_ZERO: {src} baseline on {tgt} holdout",
            batch_size=128,
            domain_vocab=static_assets.get("domain_vocab", None),
        )

        if MAKE_PLOTS:
            plt.show()

        E_ZERO_a = float(metrics_zero["RMSE_annual"])
        E_ZERO_w = float(metrics_zero["RMSE_winter"])
        E_ZERO = float(np.mean([E_ZERO_a, E_ZERO_w]))

        row = {
            "pair": pair_key,
            "source": src,
            "target": tgt,
            "E_ZERO_a": E_ZERO_a,
            "E_ZERO_w": E_ZERO_w,
            "E_ZERO": E_ZERO,
        }
        # keep any other metrics you care about
        for k in ["R2_annual", "R2_winter", "bias_annual", "bias_winter"]:
            if k in metrics_zero:
                row[k] = metrics_zero[k]

        E0_rows.append(row)
        preds_zero_by_pair[pair_key] = df_preds_zero  # optional

        print(
            f"{pair_key}: E0_a={E_ZERO_a:.3f} | E0_w={E_ZERO_w:.3f} | E0={E_ZERO:.3f}"
        )

    except Exception as e:
        if MAKE_PLOTS and fig is not None:
            plt.close(fig)
        E0_rows.append({
            "pair": pair_key,
            "source": src,
            "target": tgt,
            "error": repr(e)
        })

df_E0 = pd.DataFrame(E0_rows).sort_values(["source", "target"])
display(df_E0)
```

```python
RUN_EVAL = True

# build a quick lookup: pair_key -> (E0_a, E0_w)
# expects df_E0 with columns ["pair","E_ZERO_a","E_ZERO_w"]
E0_lookup = (df_E0.set_index("pair")[["E_ZERO_a",
                                      "E_ZERO_w"]].to_dict(orient="index"))

df_etl_adapter_by_pair = {}

for pair_key in tqdm(sorted(assets_all_by_pair.keys()),
                     desc="Evaluate (all pairs)"):
    assets_all = assets_all_by_pair[pair_key]
    models_tl_adapter = models_tl_by_pair.get(pair_key, {})
    if not assets_all:
        continue

    # per-pair output file
    out_csv = os.path.join(RESULTS_DIR, f"df_etl_adapter_{pair_key}.csv")

    if RUN_EVAL:
        rows = []
        for exp_key in tqdm(sorted(assets_all.keys()),
                            desc=f"Evaluating {pair_key}",
                            leave=False):
            run_key = f"{exp_key}__adapter"
            model = models_tl_adapter.get(run_key, None)
            if model is None:
                # checkpoint might not exist / training skipped
                continue

            assets = assets_all[exp_key]

            metrics, df_preds, _, _ = evaluate_one_model_TL(
                cfg=cfg,
                model=model,
                device=device,
                tl_assets_for_key=assets,
                ax=None,
                title=None,
                batch_size=128,
                domain_vocab=assets.get("domain_vocab", None),
                show_plot=False,
            )
            metrics["exp_key"] = exp_key
            rows.append(metrics)

        df_etl = pd.DataFrame(rows).set_index("exp_key").sort_index()

        # attach correct E_ZERO for THIS pair
        E0_a = E0_lookup.get(pair_key, {}).get("E_ZERO_a", float("nan"))
        E0_w = E0_lookup.get(pair_key, {}).get("E_ZERO_w", float("nan"))

        df_etl["pair"] = pair_key
        df_etl["source"] = pair_key.split("_to_")[0]
        df_etl["target"] = pair_key.split("_to_")[1]

        df_etl["E_ZERO_RMSE_annual"] = E0_a
        df_etl["Delta_vs_ZERO_annual"] = df_etl["RMSE_annual"] - E0_a

        df_etl["E_ZERO_RMSE_winter"] = E0_w
        df_etl["Delta_vs_ZERO_winter"] = df_etl["RMSE_winter"] - E0_w

        df_etl.to_csv(out_csv)
        df_etl_adapter_by_pair[pair_key] = df_etl

    else:
        if os.path.exists(out_csv):
            df_etl_adapter_by_pair[pair_key] = pd.read_csv(out_csv,
                                                           index_col="exp_key")
        else:
            print(f"Missing cached eval CSV for {pair_key}: {out_csv}")

# optional: one big table for all pairs
if df_etl_adapter_by_pair:
    df_etl_adapter_all = pd.concat(df_etl_adapter_by_pair.values(), axis=0)
    display(df_etl_adapter_all.head())
else:
    df_etl_adapter_all = pd.DataFrame()

df_etl_adapter_all
```

```python
# exp_key now has no _M... part
_re_budget = re.compile(r"_G(?P<G>\d+)_Y(?P<Y>\d+)_seed(?P<seed>\d+)")


def parse_budget(s: str):
    m = _re_budget.search(str(s))
    if not m:
        return {"G": np.nan, "Y": np.nan, "seed": np.nan}
    return {k: int(v) for k, v in m.groupdict().items()}


# keep exp_key as a column (don’t rely on rename quirks)
df_etl_adapter2 = df_etl_adapter.reset_index()  # creates column "exp_key"
meta = df_etl_adapter2["exp_key"].apply(parse_budget).apply(pd.Series)
df_etl_adapter2 = pd.concat([df_etl_adapter2, meta], axis=1)

# drop non-budget rows
df_etl_adapter2 = df_etl_adapter2.dropna(subset=["G", "Y", "seed"]).copy()

# aggregate across seeds per (G,Y)
agg = df_etl_adapter2.groupby(["G", "Y"]).agg(
    # ---- Annual ----
    RMSE_a_med=("RMSE_annual", "median"),
    RMSE_a_p10=("RMSE_annual", lambda x: np.quantile(x, 0.10)),
    RMSE_a_p90=("RMSE_annual", lambda x: np.quantile(x, 0.90)),

    # ---- Winter ----
    RMSE_w_med=("RMSE_winter", "median"),
    RMSE_w_p10=("RMSE_winter", lambda x: np.quantile(x, 0.10)),
    RMSE_w_p90=("RMSE_winter", lambda x: np.quantile(x, 0.90)),
    n=("RMSE_annual", "size"),
).reset_index().sort_values(["G", "Y"])

display(agg)
```

### Compute E_SCRATCH:
“no-transfer” (from-scratch) baseline trained on the same small ISL monitoring subset and evaluated on the same fixed ISL holdout

```python
RUN_SCRATCH = False

if RUN_SCRATCH:
    _re_budget = re.compile(
        r"_G(?P<G>\d+)_Y(?P<Y>\d+)_M(?P<M>\d+)_seed(?P<seed>\d+)")

    def parse_budget(exp_key: str):
        m = _re_budget.search(str(exp_key))  # works even with ..._FT_ft.joblib
        if not m:
            return {"G": np.nan, "Y": np.nan, "M": np.nan, "seed": np.nan}
        return {k: int(v) for k, v in m.groupdict().items()}

    def within_assets_from_tl_assets(tl_assets_for_key: dict):
        return {
            "ds_train": tl_assets_for_key["ds_finetune"],
            "ds_test": tl_assets_for_key["ds_test"],
            "train_idx": tl_assets_for_key["finetune_train_idx"],
            "val_idx": tl_assets_for_key["finetune_val_idx"],
        }

    models_within = {}
    infos_within = {}
    rows = []

    exp_keys = sorted(assets_all.keys())
    pbar = tqdm(exp_keys,
                desc="Training+Evaluating E_SCRATCH",
                dynamic_ncols=True)

    for exp_key in pbar:
        meta = parse_budget(exp_key)
        if not np.isnan(meta["G"]):
            pbar.set_postfix(meta)

        w_assets = within_assets_from_tl_assets(assets_all[exp_key])

        model_w, path_w, info_w = train_or_load_one_within_region(
            cfg=cfg,
            key=exp_key,
            lstm_assets=w_assets,
            best_params=params_by_key["06_ISL"],
            device=device,
            models_dir=MODELS_DIR,
            prefix="lstm_within_ISL",
            train_flag=False,
            force_retrain=False,
            epochs=150,
            batch_size_train=64,
            batch_size_val=128,
            batch_size_test=128,
            date="2026-02-27",
            verbose=False)

        models_within[exp_key] = model_w
        infos_within[exp_key] = {"model_path": path_w, **(info_w or {})}

        # ---- Evaluate ----
        met_w, df_w = model_w.evaluate_with_preds(
            device,
            info_w["test_dl"],
            info_w["ds_test"],
        )

        # ---- Seasonal metrics (annual + winter) ----
        scores_annual, scores_winter = compute_seasonal_scores(
            df_w, target_col="target", pred_col="pred")

        rows.append({
            "exp_key":
            exp_key,

            # scratch annual/winter
            "RMSE_SCRATCH_annual":
            float(scores_annual["rmse"]),
            "RMSE_SCRATCH_winter":
            float(scores_winter["rmse"]),
            "R2_SCRATCH_annual":
            float(scores_annual["R2"]),
            "R2_SCRATCH_winter":
            float(scores_winter["R2"]),
            "Bias_SCRATCH_annual":
            float(scores_annual["Bias"]),
            "Bias_SCRATCH_winter":
            float(scores_winter["Bias"]),
            "n_annual":
            int(scores_annual.get("n", np.nan)) if isinstance(
                scores_annual, dict) else np.nan,
            "n_winter":
            int(scores_winter.get("n", np.nan)) if isinstance(
                scores_winter, dict) else np.nan,
            **meta,
            "model_path":
            path_w,
        })

    df_scratch = pd.DataFrame(rows).set_index("exp_key").sort_index()

    # save
    df_scratch.to_csv(
        f"results/CH_ISL_experiment_{EXPERIMENT_NUMBER}/E_SCRATCH.csv")
    display(df_scratch.head())
else:
    df_scratch = None
```

## Evaluate experiments:

### Build one unified results dataframe (with G/Y/M/seed parsed):

```python
# 1) Parse G/Y/seed from *any* exp_key string (even if it has suffixes)
_re_budget = re.compile(r"_G(?P<G>\d+)_Y(?P<Y>\d+)_seed(?P<seed>\d+)")


def parse_budget(s: str):
    m = _re_budget.search(str(s))
    if not m:
        return {"G": np.nan, "Y": np.nan, "seed": np.nan}
    return {k: int(v) for k, v in m.groupdict().items()}


# 2) Start from TL results (has RMSE_annual/RMSE_winter etc.)
df = df_etl_adapter.copy()

# 3) Optionally join scratch results (if provided)
scratch_cols = [
    "RMSE_SCRATCH_annual",
    "RMSE_SCRATCH_winter",
    "R2_SCRATCH_annual",
    "R2_SCRATCH_winter",
    "Bias_SCRATCH_annual",
    "Bias_SCRATCH_winter",
]

df_scratch = None
if "df_scratch" in globals() and df_scratch is not None:
    scratch_cols_present = [c for c in scratch_cols if c in df_scratch.columns]
    if len(scratch_cols_present) > 0:
        # prefer inner join to keep only exp_keys available in both
        df = df.join(df_scratch[scratch_cols_present], how="inner")
    else:
        print(
            "df_scratch provided but none of the expected scratch columns are present; skipping join."
        )
else:
    df_scratch = None  # ensure name exists
    print("df_scratch not provided; proceeding without scratch metrics.")

# 4) Add parsed budget columns
meta = df.index.to_series().apply(parse_budget).apply(pd.Series)
df = pd.concat([meta, df], axis=1)

# 5) Drop any rows that didn’t parse (e.g. FULLPOOL keys)
df = df.dropna(subset=["G", "Y", "seed"]).copy()

# 6) Derived quantities for plotting (no M anymore)
df["Effort"] = df["G"] * df["Y"]

# Baselines (already computed earlier in notebook)
df["E_ZERO_a"] = float(E_ZERO_a)
df["E_ZERO_w"] = float(E_ZERO_w)
df["E_ZERO"] = float(E_ZERO)  # mean(annual,winter)

df["E_FULL_a"] = float(E_FULL_a)
df["E_FULL_w"] = float(E_FULL_w)
df["E_FULL"] = float(E_FULL)  # mean(annual,winter)

# --- TL combined metrics ---
df["RMSE_TL"] = 0.5 * (df["RMSE_annual"] + df["RMSE_winter"])

# --- Scratch-derived metrics only if scratch columns exist ---
has_scratch = (("RMSE_SCRATCH_annual" in df.columns)
               and ("RMSE_SCRATCH_winter" in df.columns))

if has_scratch:
    df["RMSE_SCRATCH"] = 0.5 * (df["RMSE_SCRATCH_annual"] +
                                df["RMSE_SCRATCH_winter"])

    # Transfer gain (positive = TL better)
    df["Transfer_gain_annual"] = df["RMSE_SCRATCH_annual"] - df["RMSE_annual"]
    df["Transfer_gain_winter"] = df["RMSE_SCRATCH_winter"] - df["RMSE_winter"]
    df["Transfer_gain"] = df["RMSE_SCRATCH"] - df["RMSE_TL"]
else:
    print(
        "Scratch RMSE columns missing; skipping RMSE_SCRATCH and Transfer_gain columns."
    )

# Deltas vs ZERO (annual/winter + combined)
df["Delta_vs_ZERO_annual"] = df["RMSE_annual"] - df["E_ZERO_a"]
df["Delta_vs_ZERO_winter"] = df["RMSE_winter"] - df["E_ZERO_w"]
df["Delta_vs_ZERO"] = df["RMSE_TL"] - df["E_ZERO"]

# Recovery (0 = ZERO, 1 = FULL)
den = (df["E_ZERO"] - df["E_FULL"])
df["Recovery_TL"] = (df["E_ZERO"] - df["RMSE_TL"]) / den
df["Recovery_TL"] = df["Recovery_TL"].clip(-0.25, 1.25)

df.head()
```
