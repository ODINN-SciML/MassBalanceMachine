### Fixed glacier hold-out split (spatial generalization):
#### Calculate directional “shift as seen from the source model

```python
from sklearn.preprocessing import StandardScaler

# Fit one global scaler across CH, NOR, ISL
# collect all feature rows across all pair results
X_global_list = []

for key, res in all_pair_res.items():
    # source side
    X_train = to_feature_matrix(res["df_train"], FEATURE_COLS, dropna=True)
    if X_train.size > 0:
        X_global_list.append(X_train)

    # target side
    X_test = to_feature_matrix(res["df_test"], FEATURE_COLS, dropna=True)
    if X_test.size > 0:
        X_global_list.append(X_test)

X_global = np.vstack(X_global_list)

global_scaler = StandardScaler()
global_scaler.fit(X_global)

print("Global scaler fitted on:", X_global.shape)

print(global_scaler.mean_)
print(global_scaler.scale_)
```

```python
pair_splits = {}

pbar = tqdm(list(all_pair_res.items()), desc="Pool/holdout + distances")
for key, res in pbar:
    src, tgt = key.split("_to_")
    pbar.set_postfix_str(key)

    # source and target frames
    df_src = res["df_train"].copy()

    df_tgt = res["df_test"].copy()
    df_tgt = df_tgt[df_tgt["SOURCE_CODE"] == tgt].copy()  # safety

    if len(df_src) == 0 or len(df_tgt) == 0:
        pair_splits[key] = {
            "source": src,
            "target": tgt,
            "error": "empty src or tgt",
        }
        continue

    # cluster-stratified glacier split on target
    df_pool, df_holdout, holdout_glaciers, pool_glaciers, gfeat, split_summary = (
        holdout_split_cluster_stratified(
            df_tgt,
            holdout_frac=0.30,
            seed=cfg.seed,
            n_clusters=6,
        ))

    # compute distances in globally standardized space
    dist = compute_topoclimatic_distances(
        df_src=df_src,
        df_pool=df_pool,
        df_holdout=df_holdout,
        feature_cols=FEATURE_COLS,
        scaler=global_scaler,
        seed=cfg.seed,
        energy_max_samples=4000,
    )

    pair_splits[key] = {
        "source": src,
        "target": tgt,
        "df_src": df_src,
        "df_target": df_tgt,
        "df_pool": df_pool,
        "df_holdout": df_holdout,
        "pool_glaciers": pool_glaciers,
        "holdout_glaciers": holdout_glaciers,
        "gfeat": gfeat,
        "split_summary": split_summary,
        "distances": dist,
    }

    print(f"{key}: "
          f"D_energy(src, tgt_all)={dist['D_energy_src_tgt_all']:.3f} | "
          f"D_wass(src, tgt_all)={dist['D_wass_src_tgt_all']:.3f} | "
          f"D_energy(src, hold)={dist['D_energy_src_holdout']:.3f} | "
          f"D_energy(pool, hold)={dist['D_energy_pool_holdout']:.3f} | ")
```

### Build static assets once (source + target holdout)

```python
def split_signature(glaciers):
    g = sorted(map(str, glaciers))
    return hashlib.md5((",".join(g)).encode("utf-8")).hexdigest()[:10]


for key, d in pair_splits.items():
    if "error" in d:
        continue
    print(key, "holdout_sig:", split_signature(d["holdout_glaciers"]),
          "n_hold:", len(d["holdout_glaciers"]))
```

```python
static_assets_by_pair = {}
errors_static_assets = {}

pbar = tqdm(list(pair_splits.items()), desc="Building static TL assets")
for key, d in pbar:
    if "error" in d:
        errors_static_assets[key] = d["error"]
        continue

    src = d["source"]
    tgt = d["target"]
    pbar.set_postfix_str(f"{src}->{tgt}")

    res_xreg = all_pair_res[key]
    holdout_glaciers = d["holdout_glaciers"]

    sig = split_signature(holdout_glaciers)
    cache_dir_pair = CACHE_DIR / key / f"holdout_{sig}"
    cache_dir_pair.mkdir(parents=True, exist_ok=True)

    def _build(force_recompute: bool):
        return build_static_tl_assets_source_and_holdout(
            cfg=cfg,
            res_xreg=res_xreg,
            target_code=tgt,
            source_code=src,
            holdout_glaciers=holdout_glaciers,
            MONTHLY_COLS=MONTHLY_COLS,
            STATIC_COLS=STATIC_COLS,
            cache_dir=cache_dir_pair,
            force_recompute=force_recompute,
            show_progress=True,
        )

    try:
        # first try: load if available
        static_assets = _build(force_recompute=False)
        static_assets_by_pair[key] = static_assets

    except Exception as e:
        msg = repr(e)

        # Only auto-recompute for the specific cache/lookup failure pattern
        should_retry = ("Missing SOURCE_CODE for ds key" in msg) or isinstance(
            e, KeyError)

        if should_retry:
            try:
                # rebuild cache cleanly
                static_assets = _build(force_recompute=True)
                static_assets_by_pair[key] = static_assets
            except Exception as e2:
                errors_static_assets[
                    key] = f"initial={msg} | recompute_failed={repr(e2)}"
        else:
            errors_static_assets[key] = msg

print("\nDone.")
print("Built assets for:", list(static_assets_by_pair.keys()))
if errors_static_assets:
    print("\nErrors:")
    for k, v in errors_static_assets.items():
        print(" ", k, ":", v)
```

### Monitoring subsamples:

#### Define the pools for experiments:

```python
pool_capacity_by_pair = {}
errors_pool_capacity = {}

pbar = tqdm(list(pair_splits.items()), desc="Computing pool capacity (G/Y/M)")
for key, d in pbar:
    if "error" in d:
        errors_pool_capacity[key] = d["error"]
        continue

    src = d["source"]
    tgt = d["target"]
    pbar.set_postfix_str(f"{src}->{tgt}")

    pool_glaciers = d["pool_glaciers"]

    # monthly target df (already filtered to target in earlier steps)
    df_tgt_monthly = d["df_target"]
    df_pool_monthly = df_tgt_monthly[df_tgt_monthly["GLACIER"].isin(
        pool_glaciers)].copy()

    if len(df_pool_monthly) == 0:
        errors_pool_capacity[key] = "empty df_pool_monthly"
        continue

    G_max = int(df_pool_monthly["GLACIER"].nunique())
    Y_max = int(df_pool_monthly.groupby("GLACIER")["YEAR"].nunique().max())

    # raw stake df for the target RGI region (needed for rows_per_gy)
    # Here we use TARGET_REGION_META[target]["rgi_region_id"] to get "06"/"08"/"11"
    rid = TARGET_REGION_META[tgt]["rgi_region_id"]  # "06", "08", "11"
    df_tgt_raw = dfs[rid].copy()
    df_tgt_raw = df_tgt_raw[df_tgt_raw["GLACIER"].isin(pool_glaciers)].copy()

    if len(df_tgt_raw) == 0:
        errors_pool_capacity[
            key] = f"empty raw df for rid={rid} after filtering pool_glaciers"
        continue

    rows_per_gy = df_tgt_raw.groupby(["GLACIER", "YEAR"]).size()
    M_p25 = int(rows_per_gy.quantile(0.25))
    M_p50 = int(rows_per_gy.median())
    M_p90 = int(rows_per_gy.quantile(0.90))
    M_max = int(rows_per_gy.max())
    M_min = int(rows_per_gy.min())

    pool_capacity_by_pair[key] = {
        "source": src,
        "target": tgt,
        "rid": rid,
        "G_max": G_max,
        "Y_max": Y_max,
        "M_min": M_min,
        "M_p25": M_p25,
        "M_p50": M_p50,
        "M_p90": M_p90,
        "M_max": M_max,
    }

    print("\n" + "=" * 80)
    print(f"POOL CAPACITY | {key}")
    print("G_max:", G_max)
    print("Y_max (max years on a glacier):", Y_max)
    print("Rows per glacier-year (raw format): min", M_min, "| p25", M_p25,
          "| median", M_p50, "| p90", M_p90, "| max", M_max)

# optional: tabular summary
df_pool_capacity = pd.DataFrame.from_dict(
    pool_capacity_by_pair,
    orient="index").reset_index().rename(columns={"index": "pair"})
display(df_pool_capacity)

if errors_pool_capacity:
    print("\nErrors:")
    for k, v in errors_pool_capacity.items():
        print(" ", k, ":", v)
```

```python
sanity_samples_by_pair = {}
errors_sanity_samples = {}

# choose a single (G, Y) for the sanity check
G_sanity = 5
Y_sanity = 5

pbar = tqdm(list(pair_splits.items()),
            desc="Sanity-check sampling (one budget)")
for key, d in pbar:
    if "error" in d:
        errors_sanity_samples[key] = d["error"]
        continue

    src = d["source"]
    tgt = d["target"]
    pbar.set_postfix_str(f"{src}->{tgt}")

    df_tgt_monthly = d["df_target"]
    pool_glaciers = d["pool_glaciers"]

    # pool rows (monthly) for this target
    df_pool_rows = df_tgt_monthly[df_tgt_monthly["GLACIER"].isin(
        pool_glaciers)].copy()
    if len(df_pool_rows) == 0:
        errors_sanity_samples[key] = "empty df_pool_rows"
        continue

    # cap G/Y so the sanity check doesn't crash on small targets
    G_eff = min(G_sanity, int(df_pool_rows["GLACIER"].nunique()))
    Y_eff = min(Y_sanity,
                int(df_pool_rows.groupby("GLACIER")["YEAR"].nunique().max()))

    try:
        df_ft, chosen, picked_years = sample_monitoring_subset_from_pool(
            df_pool=df_pool_rows,
            G=G_eff,
            Y=Y_eff,
            seed=(abs(hash(key)) + int(cfg.seed)) %
            (2**32 - 1),  # pair-specific reproducible
            glacier_pick_method="random",
            year_pick_method="earliest_block",
            enforce_full_Y_if_possible=True,
        )

        sanity_samples_by_pair[key] = {
            "df_ft": df_ft,
            "chosen_glaciers": chosen,
            "picked_years": picked_years,
            "G_used": G_eff,
            "Y_used": Y_eff,
        }

        print("\n" + "=" * 90)
        print(f"Sanity check | {key} | G={G_eff}, Y={Y_eff}")
        print("Chosen glaciers:", sorted(chosen))
        print("Years per glacier:")
        for gid in sorted(chosen):
            print(" ", gid, picked_years.get(gid))

    except Exception as e:
        errors_sanity_samples[key] = repr(e)

if errors_sanity_samples:
    print("\nErrors:")
    for k, v in errors_sanity_samples.items():
        print(" ", k, ":", v)
```

```python
# -----------------------
# Budgets & seeds (global)
# -----------------------
G_set = [1, 2, 3, 5, 10, 15, 19]  # (19 is max accross all regions)
Y_set = [1, 2, 3, 5, 10, 20, 30, 35]
BUDGETS = [{"G": int(g), "Y": int(y)} for g in G_set for y in Y_set]

N_SEEDS = 10
SEEDS = list(range(1, N_SEEDS + 1))

print("Total budget points:", len(BUDGETS))
print(pd.DataFrame(BUDGETS).sort_values(["G", "Y"]).head(12))
print("Seeds:", SEEDS)

# -----------------------------------
# Build finetune assets for ALL pairs
# -----------------------------------
assets_all_by_pair = {}  # key -> dict of assets for that pair
errors_assets_by_pair = {}  # key -> list of errors (optional)
skipped_tasks_by_pair = {}  # key -> list of skipped (G,Y,seed) (optional)

pairs_iter = tqdm(list(pair_splits.items()),
                  desc="Pairs (build finetune assets)")
for pair_key, d in pairs_iter:
    if "error" in d:
        continue

    src = d["source"]
    tgt = d["target"]
    pairs_iter.set_postfix_str(f"{src}->{tgt}")

    # Require static assets already computed
    if pair_key not in static_assets_by_pair:
        errors_assets_by_pair.setdefault(
            pair_key, []).append("missing static_assets_by_pair")
        continue

    # Pair-specific inputs
    res_xreg = all_pair_res[pair_key]
    static_assets = static_assets_by_pair[pair_key]

    # pair+split-specific cache dir (prevents "why does it recompute" issues)
    sig = split_signature(d["holdout_glaciers"])
    cache_dir_pair = CACHE_DIR / pair_key / f"holdout_{sig}"
    cache_dir_pair.mkdir(parents=True, exist_ok=True)

    # Pool rows (monthly) for sampling monitoring subsets
    pool_glaciers = d["pool_glaciers"]
    df_pool_rows = d["df_target"][d["df_target"]["GLACIER"].isin(
        pool_glaciers)].copy()

    if len(df_pool_rows) == 0:
        errors_assets_by_pair.setdefault(pair_key,
                                         []).append("empty df_pool_rows")
        continue

    # Capacity (used only to decide skipping)
    G_max = int(df_pool_rows["GLACIER"].nunique())
    Y_max = int(df_pool_rows.groupby("GLACIER")["YEAR"].nunique().max())

    # Build tasks for this pair
    TASKS = [(b, seed) for b in BUDGETS for seed in SEEDS]

    assets_all = {}
    skipped = []
    errs = []

    task_iter = tqdm(TASKS, desc=f"Assets {pair_key}", leave=False)
    for b, seed in task_iter:
        G = int(b["G"])
        Y = int(b["Y"])

        # skip budgets that can't be satisfied for this target pool
        # (avoids sample_monitoring_subset_from_pool crashing)
        if (G > G_max) or (Y > Y_max):
            skipped.append({
                "G":
                G,
                "Y":
                Y,
                "seed":
                seed,
                "reason":
                f"exceeds pool capacity (G_max={G_max}, Y_max={Y_max})"
            })
            continue

        try:
            df_ft, chosen, _ = sample_monitoring_subset_from_pool(
                df_pool=df_pool_rows,
                G=G,
                Y=Y,
                seed=seed,
                glacier_pick_method="random",
                year_pick_method="earliest_block",
                enforce_full_Y_if_possible=True,
            )

            exp_key = f"TL_{src}_to_{tgt}_G{G}_Y{Y}_seed{seed}"

            assets_one = build_budget_assets_finetune_only(
                cfg=cfg,
                res_xreg=res_xreg,
                static_assets=static_assets,
                df_ft=df_ft,
                exp_key=exp_key,
                MONTHLY_COLS=MONTHLY_COLS,
                STATIC_COLS=STATIC_COLS,
                cache_dir=cache_dir_pair,  # <-- pair-specific cache
                force_recompute=False,
                val_ratio=0.2,
                show_progress=False,
                seed=seed,
            )

            assets_all.update(assets_one)

        except Exception as e:
            errs.append({"G": G, "Y": Y, "seed": seed, "error": repr(e)})

    assets_all_by_pair[pair_key] = assets_all
    if skipped:
        skipped_tasks_by_pair[pair_key] = skipped
    if errs:
        errors_assets_by_pair[pair_key] = errs

    print(f"\n{pair_key}: built={len(assets_all)} | "
          f"skipped={len(skipped)} | errors={len(errs)} | "
          f"G_max={G_max}, Y_max={Y_max}")

print("\nDone.")
print("Pairs with assets:", list(assets_all_by_pair.keys()))
print("Total assets (all pairs):",
      sum(len(v) for v in assets_all_by_pair.values()))
```

#### Compute distribution distances:

```python
RUN_FT_DISTANCE = True

if RUN_FT_DISTANCE:
    rows = []

    budget_seed_combos = [(b, seed) for b, seed in product(BUDGETS, SEEDS)]

    pairs_iter = tqdm(
        list(pair_splits.items()),
        desc="Compute src->ft distances",
        position=0,
    )

    for pair_key, d in pairs_iter:

        if "error" in d:
            continue

        src = d["source"]
        tgt = d["target"]
        pairs_iter.set_postfix_str(f"{src}->{tgt}")

        df_src = d["df_src"]
        df_holdout = d["df_holdout"]

        pool_glaciers = d["pool_glaciers"]
        df_pool_rows = d["df_target"][d["df_target"]["GLACIER"].isin(
            pool_glaciers)].copy()

        if len(df_pool_rows) == 0:
            continue

        G_max = int(df_pool_rows["GLACIER"].nunique())
        Y_max = int(df_pool_rows.groupby("GLACIER")["YEAR"].nunique().max())

        budget_iter = tqdm(
            budget_seed_combos,
            desc=f"Budgets {src}->{tgt}",
            total=len(budget_seed_combos),
            leave=False,
            position=1,
        )

        for b, seed in budget_iter:
            G = int(b["G"])
            Y = int(b["Y"])

            budget_iter.set_postfix_str(f"G={G}, Y={Y}, seed={seed}")

            if (G > G_max) or (Y > Y_max):
                continue

            df_ft, chosen, _ = sample_monitoring_subset_from_pool(
                df_pool=df_pool_rows,
                G=G,
                Y=Y,
                seed=seed,
                glacier_pick_method="random",
                year_pick_method="earliest_block",
                enforce_full_Y_if_possible=True,
            )

            dist = compute_topoclimatic_distances_sets(
                df_src=df_src,
                df_ft=df_ft,
                df_holdout=df_holdout,
                feature_cols=FEATURE_COLS,
                scaler=global_scaler,
                seed=seed,
                energy_max_samples=4000,
            )

            exp_key = f"TL_{src}_to_{tgt}_G{G}_Y{Y}_seed{seed}"

            rows.append({
                "pair": pair_key,
                "source": src,
                "target": tgt,
                "G": G,
                "Y": Y,
                "seed": seed,
                "exp_key": exp_key,
                **dist,
            })

    df_dist_src_ft = pd.DataFrame(rows)
    df_dist_src_ft.to_csv(
        os.path.join(RESULTS_DIR, "df_dist_src_ft_all_pairs.csv"),
        index=False,
    )

else:
    df_dist_src_ft = pd.read_csv(
        os.path.join(RESULTS_DIR, "df_dist_src_ft_all_pairs.csv")
    )
```

```python
RUN_E_ZERO = True
if RUN_E_ZERO:
    E0_rows = []
    preds_zero_by_pair = {}  # optional: store preds dfs

    MAKE_PLOTS = False  # set True if you want a figure per pair

    for src, tgt in tqdm(PAIRS, desc="E_ZERO for all directed pairs"):
        pair_key = f"{src}_to_{tgt}"

        if pair_key not in static_assets_by_pair:
            E0_rows.append({
                "pair": pair_key,
                "source": src,
                "target": tgt,
                "error": "missing static_assets"
            })
            continue
        if src not in baseline_models:
            E0_rows.append({
                "pair": pair_key,
                "source": src,
                "target": tgt,
                "error": "missing baseline_model"
            })
            continue

        static_assets = static_assets_by_pair[pair_key]
        model_src = baseline_models[src]

        # optional plot axis
        ax = None
        fig = None
        if MAKE_PLOTS:
            fig, ax = plt.subplots(1, 1, figsize=(6, 6))

        try:
            metrics_zero, df_preds_zero, _, _ = evaluate_one_model_TL(
                cfg=cfg,
                model=model_src,
                device=device,
                tl_assets_for_key=static_assets,
                ax=ax,
                title=f"E_ZERO: {src} baseline on {tgt} holdout",
                batch_size=128,
                domain_vocab=static_assets.get("domain_vocab", None),
            )

            if MAKE_PLOTS:
                plt.show()

            E_ZERO_a = float(metrics_zero["RMSE_annual"])
            E_ZERO_w = float(metrics_zero["RMSE_winter"])
            E_ZERO = float(np.mean([E_ZERO_a, E_ZERO_w]))

            row = {
                "pair": pair_key,
                "source": src,
                "target": tgt,
                "E_ZERO_a": E_ZERO_a,
                "E_ZERO_w": E_ZERO_w,
                "E_ZERO": E_ZERO,
            }
            # keep any other metrics you care about
            for k in ["R2_annual", "R2_winter", "bias_annual", "bias_winter"]:
                if k in metrics_zero:
                    row[k] = metrics_zero[k]

            E0_rows.append(row)
            preds_zero_by_pair[pair_key] = df_preds_zero  # optional

            print(
                f"{pair_key}: E0_a={E_ZERO_a:.3f} | E0_w={E_ZERO_w:.3f} | E0={E_ZERO:.3f}"
            )

        except Exception as e:
            if MAKE_PLOTS and fig is not None:
                plt.close(fig)
            E0_rows.append({
                "pair": pair_key,
                "source": src,
                "target": tgt,
                "error": repr(e)
            })

    df_E0 = pd.DataFrame(E0_rows).sort_values(["source", "target"])

    # Save final results to CSV
    final_results_csv = RESULTS_DIR / "df_E0_all_pairs.csv"
    df_E0.to_csv(final_results_csv, index=False)

else:
    df_E0 = pd.read_csv(os.path.join(RESULTS_DIR, "df_E0_all_pairs.csv"))
display(df_E0)
```

```python
RUN_EVAL = False

# build a quick lookup: pair_key -> (E0_a, E0_w)
# expects df_E0 with columns ["pair","E_ZERO_a","E_ZERO_w"]
E0_lookup = (df_E0.set_index("pair")[["E_ZERO_a",
                                      "E_ZERO_w"]].to_dict(orient="index"))

df_etl_adapter_by_pair = {}

for pair_key in tqdm(sorted(assets_all_by_pair.keys()),
                     desc="Evaluate (all pairs)"):
    assets_all = assets_all_by_pair[pair_key]
    models_tl_adapter = models_tl_by_pair.get(pair_key, {})
    if not assets_all:
        continue

    # per-pair output file
    out_csv = os.path.join(RESULTS_DIR, f"df_etl_adapter_{pair_key}.csv")

    if RUN_EVAL:
        rows = []
        for exp_key in tqdm(sorted(assets_all.keys()),
                            desc=f"Evaluating {pair_key}",
                            leave=False):
            run_key = f"{exp_key}__adapter"
            model = models_tl_adapter.get(run_key, None)
            if model is None:
                # checkpoint might not exist / training skipped
                continue

            assets = assets_all[exp_key]

            metrics, df_preds, _, _ = evaluate_one_model_TL(
                cfg=cfg,
                model=model,
                device=device,
                tl_assets_for_key=assets,
                ax=None,
                title=None,
                batch_size=128,
                domain_vocab=assets.get("domain_vocab", None),
                show_plot=False,
            )
            metrics["exp_key"] = exp_key
            rows.append(metrics)

        df_etl = pd.DataFrame(rows).set_index("exp_key").sort_index()

        # attach correct E_ZERO for THIS pair
        E0_a = E0_lookup.get(pair_key, {}).get("E_ZERO_a", float("nan"))
        E0_w = E0_lookup.get(pair_key, {}).get("E_ZERO_w", float("nan"))

        df_etl["pair"] = pair_key
        df_etl["source"] = pair_key.split("_to_")[0]
        df_etl["target"] = pair_key.split("_to_")[1]

        df_etl["E_ZERO_RMSE_annual"] = E0_a
        df_etl["Delta_vs_ZERO_annual"] = df_etl["RMSE_annual"] - E0_a

        df_etl["E_ZERO_RMSE_winter"] = E0_w
        df_etl["Delta_vs_ZERO_winter"] = df_etl["RMSE_winter"] - E0_w

        df_etl.to_csv(out_csv)
        df_etl_adapter_by_pair[pair_key] = df_etl

    else:
        if os.path.exists(out_csv):
            df_etl_adapter_by_pair[pair_key] = pd.read_csv(out_csv,
                                                           index_col="exp_key")
        else:
            print(f"Missing cached eval CSV for {pair_key}: {out_csv}")

# optional: one big table for all pairs
if df_etl_adapter_by_pair:
    df_etl_adapter_all = pd.concat(df_etl_adapter_by_pair.values(), axis=0)
else:
    df_etl_adapter_all = pd.DataFrame()

display(df_etl_adapter_all.head())
```

```python
import re
import numpy as np
import pandas as pd
from tqdm.auto import tqdm

RUN_SCRATCH = True

if RUN_SCRATCH:
    _re_budget = re.compile(
        r"_G(?P<G>\d+)_Y(?P<Y>\d+)_seed(?P<seed>\d+)"
    )

    def parse_budget(exp_key: str):
        m = _re_budget.search(str(exp_key))
        if not m:
            return {"G": np.nan, "Y": np.nan, "seed": np.nan}
        return {k: int(v) for k, v in m.groupdict().items()}

    def within_assets_from_tl_assets(tl_assets_for_key: dict):
        """
        Train only on the fine-tuning subset and evaluate on the holdout test set.
        """
        return {
            "ds_train": tl_assets_for_key["ds_finetune"],
            "ds_test": tl_assets_for_key["ds_test"],
            "train_idx": tl_assets_for_key["finetune_train_idx"],
            "val_idx": tl_assets_for_key["finetune_val_idx"],
        }

    def infer_src_tgt_from_exp_key(exp_key: str):
        m = re.search(r"TL_(?P<src>.+?)_to_(?P<tgt>.+?)_G\d+_Y\d+_seed\d+", str(exp_key))
        if not m:
            return None, None
        return m.group("src"), m.group("tgt")

    models_within = {}
    infos_within = {}
    rows = []

    exp_keys = sorted(assets_all.keys())
    pbar = tqdm(
        exp_keys,
        desc="Training + evaluating scratch models",
        dynamic_ncols=True,
    )

    for exp_key in pbar:
        meta = parse_budget(exp_key)
        if not np.isnan(meta["G"]):
            pbar.set_postfix(
                G=int(meta["G"]),
                Y=int(meta["Y"]),
                seed=int(meta["seed"]),
            )

        tl_assets_for_key = assets_all[exp_key]
        w_assets = within_assets_from_tl_assets(tl_assets_for_key)

        # Prefer metadata from assets if available
        src = tl_assets_for_key.get("source_code", None)
        tgt = tl_assets_for_key.get("target_code", None)

        if src is None or tgt is None:
            src2, tgt2 = infer_src_tgt_from_exp_key(exp_key)
            src = src if src is not None else src2
            tgt = tgt if tgt is not None else tgt2

        if src is None:
            raise ValueError(f"Could not infer source region for exp_key={exp_key}")

        best_params = params_by_key[REGION_CODE_MAP[src]]

        model_w, path_w, info_w = train_or_load_one_within_region_monitoring(
            cfg=cfg,
            key=exp_key,
            lstm_assets=w_assets,
            best_params=best_params,
            device=device,
            models_dir=MODELS_DIR,
            prefix="lstm_within",
            train_flag=True,
            force_retrain=False,
            epochs=150,
            batch_size_train=64,
            batch_size_val=128,
            batch_size_test=128,
            date=None,
            verbose=False,
        )

        models_within[exp_key] = model_w
        infos_within[exp_key] = {"model_path": path_w, **(info_w or {})}

        # Evaluate on holdout
        met_w, df_w = model_w.evaluate_with_preds(
            device,
            info_w["test_dl"],
            info_w["ds_test"],
        )

        scores_annual, scores_winter = compute_seasonal_scores(
            df_w,
            target_col="target",
            pred_col="pred",
        )

        rows.append({
            "exp_key": exp_key,
            "source": src,
            "target": tgt,

            "RMSE_SCRATCH_annual": float(scores_annual["rmse"]),
            "RMSE_SCRATCH_winter": float(scores_winter["rmse"]),
            "R2_SCRATCH_annual": float(scores_annual["R2"]),
            "R2_SCRATCH_winter": float(scores_winter["R2"]),
            "Bias_SCRATCH_annual": float(scores_annual["Bias"]),
            "Bias_SCRATCH_winter": float(scores_winter["Bias"]),
            "n_annual": int(scores_annual.get("n", np.nan)),
            "n_winter": int(scores_winter.get("n", np.nan)),

            **meta,
            "model_path": path_w,
        })

    df_scratch = pd.DataFrame(rows).set_index("exp_key").sort_index()

    final_results_csv = RESULTS_DIR / "df_ESCRATCH_all_pairs.csv"
    df_scratch.to_csv(final_results_csv)

    display(df_scratch.head())

else:
    df_scratch = None
```

## Evaluate experiments:

### Build one unified results dataframe (with G/Y/M/seed parsed):

```python
# ----------------------------
# 1) Parse G/Y/seed from exp_key
# ----------------------------
_re_budget = re.compile(r"_G(?P<G>\d+)_Y(?P<Y>\d+)_seed(?P<seed>\d+)")


def parse_budget(s: str):
    m = _re_budget.search(str(s))
    if not m:
        return {"G": np.nan, "Y": np.nan, "seed": np.nan}
    return {k: int(v) for k, v in m.groupdict().items()}


# ----------------------------
# 2) Start from TL experiment results
# ----------------------------
df = df_etl_adapter_all.copy()

if "exp_key" not in df.columns:
    df = df.reset_index()

# ----------------------------
# 3) Add parsed budget columns
# ----------------------------
meta = df["exp_key"].apply(parse_budget).apply(pd.Series)
df = pd.concat([df, meta], axis=1)

# Drop non-budget rows (e.g. FULLPOOL if present)
df = df.dropna(subset=["G", "Y", "seed"]).copy()

df["G"] = df["G"].astype(int)
df["Y"] = df["Y"].astype(int)
df["seed"] = df["seed"].astype(int)

# ----------------------------
# 4) Merge pair-specific E_ZERO
# ----------------------------
df = df.merge(
    df_E0[["pair", "E_ZERO_a", "E_ZERO_w", "E_ZERO"]],
    on="pair",
    how="left",
)

# ----------------------------
# 5) Merge pair-specific E_FULL
# ----------------------------
df = df.merge(
    df_EFULL[["pair", "E_FULL_a", "E_FULL_w", "E_FULL"]],
    on="pair",
    how="left",
)

# ----------------------------
# 6) Build a flat dataframe of pair-level topoclimatic distances
# ----------------------------
pair_distance_rows = []
for pair_key, d in pair_splits.items():
    if "error" in d:
        continue

    row = {
        "pair": pair_key,
        "source": d["source"],
        "target": d["target"],
    }
    row.update(d["distances"])
    pair_distance_rows.append(row)

df_pair_dist = pd.DataFrame(pair_distance_rows)

# Merge pair-level distances
df = df.merge(
    df_pair_dist,
    on=["pair", "source", "target"],
    how="left",
)

# ----------------------------
# 7) Merge experiment-level src->ft distances
# ----------------------------
df = df.merge(
    df_dist_src_ft[[
        "exp_key",
        "n_ft",
        "D_wass_src_ft",
        "D_energy_src_ft",
    ]],
    on="exp_key",
    how="left",
)

# ----------------------------
# 8) Derived quantities
# ----------------------------
df["Effort"] = df["G"] * df["Y"]
df["RMSE_TL"] = 0.5 * (df["RMSE_annual"] + df["RMSE_winter"])

# Deltas vs ZERO
df["Delta_vs_ZERO_annual"] = df["RMSE_annual"] - df["E_ZERO_a"]
df["Delta_vs_ZERO_winter"] = df["RMSE_winter"] - df["E_ZERO_w"]
df["Delta_vs_ZERO"] = df["RMSE_TL"] - df["E_ZERO"]

df["GapClosed_annual"] = ((df["E_ZERO_a"] - df["RMSE_annual"]) /
                          (df["E_ZERO_a"] - df["E_FULL_a"]))

df["GapClosed_winter"] = ((df["E_ZERO_w"] - df["RMSE_winter"]) /
                          (df["E_ZERO_w"] - df["E_FULL_w"]))

df["GapClosed_mean"] = ((df["E_ZERO"] - df["RMSE_TL"]) /
                        (df["E_ZERO"] - df["E_FULL"]))

df["GapClosed_annual"] = df["GapClosed_annual"].clip(-0.25, 1.25)
df["GapClosed_winter"] = df["GapClosed_winter"].clip(-0.25, 1.25)
df["GapClosed_mean"] = df["GapClosed_mean"].clip(-0.25, 1.25)

# ----------------------------
# 9) Quick sanity checks
# ----------------------------
print("Missing E_ZERO:", df["E_ZERO"].isna().sum())
print("Missing E_FULL:", df["E_FULL"].isna().sum())
print("Missing pair distances:", df["D_energy_src_tgt_all"].isna().sum())
print("Missing FT distances:", df["D_energy_src_ft"].isna().sum())

# Save final merged dataframe for analysis
final_merged_csv = RESULTS_DIR / "df_final_merged_all_pairs.csv"
df.to_csv(final_merged_csv, index=False)
df.head()
```

### Distance plots:

```python
def plot_recovery_by_distance_bins(
    df,
    metric_col="GapClosed_annual",
    dist_col="D_energy_src_tgt_all",
    g_col="G",
    y_col="Y",
    pair_col="pair",
    q=3,
    gmin=1,
    ymin=1,
    bin_labels=None,
    levels_filled=None,
    levels_lines=None,
    cmap="viridis",
    vmin=0.0,
    vmax=1.0,
    figsize=None,
    sharey=True,
    show_pair_labels=True,
    show_bounds_in_title=True,
    pair_label_transform=lambda p: p.replace("_to_", " → "),
    xlabel="Years monitored (Y)",
    ylabel="Glaciers monitored (G)",
    suptitle="Monitoring recovery by topoclimatic transfer distance",
    cbar_label=None,
    title_dist_symbol=r"D_E^{src\to tgt}",
    scatter_points=True,
    scatter_kwargs=None,
    text_kwargs=None,
    contour_kwargs=None,
    contourf_kwargs=None,
):
    """
    Plot mean recovery surfaces over (G, Y), grouped by quantile bins of a distance metric.

    Parameters
    ----------
    df : pd.DataFrame
        Input dataframe containing metric, distance, G, Y, and pair columns.
    metric_col : str
        Column to plot as the surface value.
    dist_col : str
        Distance column used for binning.
    g_col : str
        Column for number of glaciers monitored.
    y_col : str
        Column for number of years monitored.
    pair_col : str
        Column identifying transfer pairs.
    q : int
        Number of quantile bins.
    bin_labels : list[str] or None
        Labels for the distance bins. If None, defaults depend on q.
    levels_filled : array-like or None
        Filled contour levels. If None, uses np.linspace(vmin, vmax, 11).
    levels_lines : list[float] or None
        Contour line levels. If None, uses [0.25, 0.50, 0.75, 1.00].
    cmap : str
        Colormap for filled contours.
    vmin, vmax : float
        Color scale limits.
    figsize : tuple or None
        Figure size. If None, uses (6*q, 5).
    sharey : bool
        Whether subplots share the y-axis.
    show_pair_labels : bool
        Whether to annotate each panel with the pairs in that bin.
    show_bounds_in_title : bool
        Whether to include distance bin bounds in each panel title.
    pair_label_transform : callable
        Function to convert raw pair keys to display labels.
    xlabel, ylabel : str
        Axis labels.
    suptitle : str
        Figure super-title.
    cbar_label : str or None
        Colorbar label. If None, uses metric_col.
    title_dist_symbol : str
        MathText string for the distance symbol in titles.
    scatter_points : bool
        Whether to overlay the actual (Y, G) grid points.
    scatter_kwargs, text_kwargs, contour_kwargs, contourf_kwargs : dict or None
        Optional matplotlib styling dictionaries.

    Returns
    -------
    fig, axes : matplotlib Figure and Axes
    """
    # Filter to min G and Y
    df = df[df[g_col] >= gmin].copy()
    df = df[df[y_col] >= ymin].copy()

    required_cols = [metric_col, dist_col, g_col, y_col, pair_col]
    missing = [c for c in required_cols if c not in df.columns]
    if missing:
        raise KeyError(f"Missing required columns: {missing}")

    df_plot = df.dropna(subset=required_cols).copy()
    if df_plot.empty:
        raise ValueError("No valid rows left after dropping NaNs.")

    if bin_labels is None:
        default_labels = {
            2: ["Low distance", "High distance"],
            3: ["Low distance", "Medium distance", "High distance"],
            4: ["Very low", "Low", "High", "Very high"],
        }
        bin_labels = default_labels.get(q, [f"Bin {i+1}" for i in range(q)])

    if len(bin_labels) != q:
        raise ValueError(
            f"bin_labels must have length {q}, got {len(bin_labels)}.")

    if levels_filled is None:
        levels_filled = np.linspace(vmin, vmax, 11)
    if levels_lines is None:
        levels_lines = [0.25, 0.50, 0.75, 1.00]
    if figsize is None:
        figsize = (6 * q, 5)
    if cbar_label is None:
        cbar_label = metric_col

    if scatter_kwargs is None:
        scatter_kwargs = dict(s=16, alpha=0.45)
    if text_kwargs is None:
        text_kwargs = dict(
            x=0.03,
            y=0.97,
            ha="left",
            va="top",
            fontsize=8,
            bbox=dict(
                boxstyle="round,pad=0.25",
                facecolor="white",
                alpha=0.85,
                edgecolor="none",
            ),
        )
    if contour_kwargs is None:
        contour_kwargs = dict(colors="k", linewidths=1.0)
    if contourf_kwargs is None:
        contourf_kwargs = {}

    # Keep raw interval bins
    df_plot["dist_bin_raw"] = pd.qcut(df_plot[dist_col], q=q)

    raw_bins = df_plot["dist_bin_raw"].cat.categories
    bin_label_map = {
        raw_bin: label
        for raw_bin, label in zip(raw_bins, bin_labels)
    }
    df_plot["dist_bin"] = df_plot["dist_bin_raw"].map(bin_label_map)

    pair_bin = (df_plot[[pair_col, "dist_bin"]].drop_duplicates().sort_values(
        ["dist_bin", pair_col]))

    fig, axes = plt.subplots(1,
                             q,
                             figsize=figsize,
                             sharey=sharey,
                             squeeze=False)
    axes = axes.ravel()

    im = None

    for ax, raw_bin, label in zip(axes, raw_bins, bin_labels):
        sub = df_plot[df_plot["dist_bin_raw"] == raw_bin].copy()

        heat = (sub.groupby([g_col, y_col
                             ])[metric_col].mean().unstack(y_col).sort_index(
                                 axis=0).sort_index(axis=1))

        if heat.empty:
            ax.set_visible(False)
            continue

        y_vals = heat.columns.to_numpy(dtype=float)
        g_vals = heat.index.to_numpy(dtype=float)
        X, Y = np.meshgrid(y_vals, g_vals)
        Z = heat.to_numpy(dtype=float)

        cf = ax.contourf(
            X,
            Y,
            Z,
            levels=levels_filled,
            vmin=vmin,
            vmax=vmax,
            cmap=cmap,
            **contourf_kwargs,
        )
        im = cf

        # Only draw contour lines when there is enough variation
        zmin = np.nanmin(Z)
        zmax = np.nanmax(Z)
        valid_line_levels = [lv for lv in levels_lines if zmin <= lv <= zmax]
        if len(valid_line_levels) > 0:
            cs = ax.contour(
                X,
                Y,
                Z,
                levels=valid_line_levels,
                **contour_kwargs,
            )
            ax.clabel(cs, fmt="%.2f", inline=True, fontsize=8)

        if scatter_points:
            ax.scatter(X, Y, **scatter_kwargs)

        if show_pair_labels:
            pairs_here = sorted(pair_bin.loc[pair_bin["dist_bin"] == label,
                                             pair_col].tolist())
            pair_txt = "\n".join([pair_label_transform(p) for p in pairs_here])

            if pair_txt.strip():
                ax.text(
                    text_kwargs.pop("x"),
                    text_kwargs.pop("y"),
                    pair_txt,
                    transform=ax.transAxes,
                    **text_kwargs,
                )
                # restore x/y for next panel
                text_kwargs["x"] = 0.03
                text_kwargs["y"] = 0.97

        if show_bounds_in_title:
            lo = raw_bin.left
            hi = raw_bin.right
            ax.set_title(
                f"{label}\n${lo:.2f} < {title_dist_symbol} \\leq {hi:.2f}$")
        else:
            ax.set_title(label)

        ax.set_xlabel(xlabel)
        ax.set_xticks(y_vals)
        ax.set_yticks(g_vals)

    axes[0].set_ylabel(ylabel)

    fig.suptitle(suptitle, y=1.0)

    plt.tight_layout()
    fig.subplots_adjust(right=0.88, top=0.82)

    cax = fig.add_axes([0.90, 0.15, 0.02, 0.7])
    cbar = fig.colorbar(im, cax=cax)
    cbar.set_label(cbar_label)

    return fig, axes
```

```python
def plot_metric_by_distance_bins(
    df,
    metric_col="GapClosed_annual",
    dist_col="D_energy_src_tgt_all",
    g_col="G",
    y_col="Y",
    pair_col="pair",
    q=3,
    bin_labels=None,
    levels_filled=None,
    levels_lines=None,
    cmap="viridis",
    vmin=None,
    vmax=None,
    figsize=None,
    sharey=True,
    show_pair_labels=True,
    show_bounds_in_title=True,
    pair_label_transform=lambda p: p.replace("_to_", " → "),
    xlabel="Years monitored (Y)",
    ylabel="Glaciers monitored (G)",
    suptitle="Monitoring response by topoclimatic transfer distance",
    cbar_label=None,
    title_dist_symbol=r"D_E^{src\to tgt}",
    scatter_points=True,
    scatter_kwargs=None,
    text_kwargs=None,
    contour_kwargs=None,
    contourf_kwargs=None,
):
    required_cols = [metric_col, dist_col, g_col, y_col, pair_col]
    missing = [c for c in required_cols if c not in df.columns]
    if missing:
        raise KeyError(f"Missing required columns: {missing}")

    df_plot = df.dropna(subset=required_cols).copy()
    if df_plot.empty:
        raise ValueError("No valid rows left after dropping NaNs.")

    if bin_labels is None:
        default_labels = {
            2: ["Low distance", "High distance"],
            3: ["Low distance", "Medium distance", "High distance"],
            4: ["Very low", "Low", "High", "Very high"],
        }
        bin_labels = default_labels.get(q, [f"Bin {i+1}" for i in range(q)])

    if len(bin_labels) != q:
        raise ValueError(
            f"bin_labels must have length {q}, got {len(bin_labels)}.")

    if vmin is None:
        vmin = float(np.nanmin(df_plot[metric_col]))
    if vmax is None:
        vmax = float(np.nanmax(df_plot[metric_col]))

    if levels_filled is None:
        levels_filled = np.linspace(vmin, vmax, 11)
    if levels_lines is None:
        levels_lines = np.linspace(vmin, vmax, 5)[1:-1]
    if figsize is None:
        figsize = (6 * q, 5)
    if cbar_label is None:
        cbar_label = metric_col

    if scatter_kwargs is None:
        scatter_kwargs = dict(s=16, alpha=0.45)
    if text_kwargs is None:
        text_kwargs = dict(
            ha="left",
            va="top",
            fontsize=8,
            bbox=dict(
                boxstyle="round,pad=0.25",
                facecolor="white",
                alpha=0.85,
                edgecolor="none",
            ),
        )
    if contour_kwargs is None:
        contour_kwargs = dict(colors="k", linewidths=1.0)
    if contourf_kwargs is None:
        contourf_kwargs = {}

    df_plot["dist_bin_raw"] = pd.qcut(df_plot[dist_col], q=q)

    raw_bins = df_plot["dist_bin_raw"].cat.categories
    bin_label_map = {
        raw_bin: label
        for raw_bin, label in zip(raw_bins, bin_labels)
    }
    df_plot["dist_bin"] = df_plot["dist_bin_raw"].map(bin_label_map)

    pair_bin = (df_plot[[pair_col, "dist_bin"]].drop_duplicates().sort_values(
        ["dist_bin", pair_col]))

    fig, axes = plt.subplots(1,
                             q,
                             figsize=figsize,
                             sharey=sharey,
                             squeeze=False)
    axes = axes.ravel()

    im = None

    for ax, raw_bin, label in zip(axes, raw_bins, bin_labels):
        sub = df_plot[df_plot["dist_bin_raw"] == raw_bin].copy()

        heat = (sub.groupby([g_col, y_col
                             ])[metric_col].mean().unstack(y_col).sort_index(
                                 axis=0).sort_index(axis=1))

        if heat.empty:
            ax.set_visible(False)
            continue

        y_vals = heat.columns.to_numpy(dtype=float)
        g_vals = heat.index.to_numpy(dtype=float)
        X, Y = np.meshgrid(y_vals, g_vals)
        Z = heat.to_numpy(dtype=float)

        cf = ax.contourf(
            X,
            Y,
            Z,
            levels=levels_filled,
            vmin=vmin,
            vmax=vmax,
            cmap=cmap,
            **contourf_kwargs,
        )
        im = cf

        zmin = np.nanmin(Z)
        zmax = np.nanmax(Z)
        valid_line_levels = [lv for lv in levels_lines if zmin <= lv <= zmax]
        if valid_line_levels:
            cs = ax.contour(X,
                            Y,
                            Z,
                            levels=valid_line_levels,
                            **contour_kwargs)
            ax.clabel(cs, fmt="%.2f", inline=True, fontsize=8)

        if scatter_points:
            ax.scatter(X, Y, **scatter_kwargs)

        if show_pair_labels:
            pairs_here = sorted(pair_bin.loc[pair_bin["dist_bin"] == label,
                                             pair_col].tolist())
            pair_txt = "\n".join([pair_label_transform(p) for p in pairs_here])

            if pair_txt.strip():
                ax.text(
                    0.03,
                    0.97,
                    pair_txt,
                    transform=ax.transAxes,
                    **text_kwargs,
                )

        if show_bounds_in_title:
            lo = raw_bin.left
            hi = raw_bin.right
            ax.set_title(
                f"{label}\n${lo:.2f} < {title_dist_symbol} \\leq {hi:.2f}$")
        else:
            ax.set_title(label)

        ax.set_xlabel(xlabel)
        ax.set_xticks(y_vals)
        ax.set_yticks(g_vals)

    axes[0].set_ylabel(ylabel)

    fig.suptitle(suptitle, y=1.0)

    plt.tight_layout()
    fig.subplots_adjust(right=0.88, top=0.82)

    cax = fig.add_axes([0.90, 0.15, 0.02, 0.7])
    cbar = fig.colorbar(im, cax=cax)
    cbar.set_label(cbar_label)

    return fig, axes
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


def plot_distance_vs_metric_with_std(
        df,
        dist_col="D_wass_src_ft",
        metric_col="RMSE_annual",
        group_cols=("G", "Y"),
        pair_col=None,
        figsize=(7, 5),
        xlabel=None,
        ylabel=None,
        title=None,
        annotate=True,
        color_by="G",  # "G", "Y", or None
):
    """
    Scatter plot of distance vs performance, aggregated over group_cols,
    with vertical error bars showing std(metric).

    Parameters
    ----------
    df : pd.DataFrame
    dist_col : str
        Distance column on x-axis.
    metric_col : str
        Metric column on y-axis.
    group_cols : tuple[str]
        Columns used to aggregate points, typically ("G", "Y").
    pair_col : str or None
        Optional extra grouping column, e.g. "pair".
        If provided, one point per (pair, G, Y).
    figsize : tuple
    xlabel, ylabel, title : str or None
    annotate : bool
        Whether to annotate points with "(G,Y)".
    color_by : str or None
        Column in summary table used for coloring points, typically "G" or "Y".
    """
    cols = [dist_col, metric_col, *group_cols]
    if pair_col is not None:
        cols.append(pair_col)

    df_plot = df.dropna(subset=cols).copy()

    gb_cols = list(group_cols)
    if pair_col is not None:
        gb_cols = [pair_col] + gb_cols

    summary = (df_plot.groupby(gb_cols).agg(
        dist_mean=(dist_col, "mean"),
        dist_std=(dist_col, "std"),
        metric_mean=(metric_col, "mean"),
        metric_std=(metric_col, "std"),
        n=(metric_col, "size"),
    ).reset_index())

    fig, ax = plt.subplots(figsize=figsize)

    if color_by is not None and color_by in summary.columns:
        sc = ax.scatter(
            summary["dist_mean"],
            summary["metric_mean"],
            c=summary[color_by],
            s=60,
            alpha=0.85,
        )
        cbar = fig.colorbar(sc, ax=ax)
        cbar.set_label(color_by)
    else:
        ax.scatter(
            summary["dist_mean"],
            summary["metric_mean"],
            s=60,
            alpha=0.85,
        )

    ax.errorbar(
        summary["dist_mean"],
        summary["metric_mean"],
        yerr=summary["metric_std"],
        fmt="none",
        capsize=3,
        alpha=0.7,
    )

    # optional x-errorbars too, if useful
    # ax.errorbar(
    #     summary["dist_mean"],
    #     summary["metric_mean"],
    #     xerr=summary["dist_std"],
    #     fmt="none",
    #     capsize=3,
    #     alpha=0.4,
    # )

    if annotate:
        for _, r in summary.iterrows():
            label = f"G={int(r['G'])}, Y={int(r['Y'])}"
            if pair_col is not None:
                label = f"{r[pair_col].replace('_to_', '→')}\n{label}"
            ax.annotate(
                label,
                (r["dist_mean"], r["metric_mean"]),
                xytext=(4, 4),
                textcoords="offset points",
                fontsize=8,
            )

    if xlabel is None:
        xlabel = dist_col
    if ylabel is None:
        ylabel = metric_col
    if title is None:
        title = f"{metric_col} vs {dist_col}"

    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_title(title)

    plt.tight_layout()
    return fig, ax, summary
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


def plot_distance_vs_metric_with_std_zoomed(
    df,
    dist_col="D_wass_src_ft",
    metric_col="RMSE_annual",
    group_cols=("G", "Y"),
    pair_col="pair",
    x_sections=None,
    figsize=(14, 10),
    xlabel=r"$D_{wass}^{src\to ft}$",
    ylabel="RMSE (annual)",
    suptitle="Transfer-learning RMSE vs Wasserstein distance by pair",
    annotate=False,
    color_by="G",
    sharey=True,
    sharex=False,
    show_xerr=False,
):
    """
    Scatter plot of distance vs metric with std error bars, split into multiple
    x-range panels.

    Each point is aggregated over group_cols (and optionally pair_col):
      - x = mean distance
      - y = mean metric
      - yerr = std(metric)
      - optional xerr = std(distance)

    Parameters
    ----------
    df : pd.DataFrame
    dist_col : str
    metric_col : str
    group_cols : tuple[str]
    pair_col : str or None
    x_sections : list[tuple[float, float]] or None
        List of x-axis ranges [(xmin, xmax), ...].
        If None, 4 quantile-based sections are created.
    figsize : tuple
    xlabel, ylabel, suptitle : str
    annotate : bool
    color_by : str or None
    sharey, sharex : bool
    show_xerr : bool

    Returns
    -------
    fig, axes, summary : matplotlib figure, axes, and aggregated dataframe
    """
    cols = [dist_col, metric_col, *group_cols]
    if pair_col is not None:
        cols.append(pair_col)

    df_plot = df.dropna(subset=cols).copy()

    gb_cols = list(group_cols)
    if pair_col is not None:
        gb_cols = [pair_col] + gb_cols

    summary = (df_plot.groupby(gb_cols).agg(
        dist_mean=(dist_col, "mean"),
        dist_std=(dist_col, "std"),
        metric_mean=(metric_col, "mean"),
        metric_std=(metric_col, "std"),
        n=(metric_col, "size"),
    ).reset_index())

    # Default: 4 quantile-based x sections
    if x_sections is None:
        qs = summary["dist_mean"].quantile([0.0, 0.25, 0.50, 0.75,
                                            1.0]).to_numpy()
        x_sections = [(qs[i], qs[i + 1]) for i in range(4)]

    n_panels = len(x_sections)
    ncols = 2
    nrows = int(np.ceil(n_panels / ncols))

    fig, axes = plt.subplots(nrows,
                             ncols,
                             figsize=figsize,
                             sharey=sharey,
                             sharex=sharex,
                             squeeze=False)
    axes = axes.ravel()

    scatter_ref = None

    for ax, (xmin, xmax) in zip(axes, x_sections):
        # include right edge on last bin
        mask = (summary["dist_mean"] >= xmin) & (summary["dist_mean"] <= xmax)
        sub = summary.loc[mask].copy()

        if sub.empty:
            ax.set_visible(False)
            continue

        if color_by is not None and color_by in sub.columns:
            sc = ax.scatter(
                sub["dist_mean"],
                sub["metric_mean"],
                c=sub[color_by],
                s=60,
                alpha=0.85,
            )
            scatter_ref = sc
        else:
            ax.scatter(
                sub["dist_mean"],
                sub["metric_mean"],
                s=60,
                alpha=0.85,
            )

        ax.errorbar(
            sub["dist_mean"],
            sub["metric_mean"],
            yerr=sub["metric_std"],
            fmt="none",
            capsize=3,
            alpha=0.7,
        )

        if show_xerr:
            ax.errorbar(
                sub["dist_mean"],
                sub["metric_mean"],
                xerr=sub["dist_std"],
                fmt="none",
                capsize=3,
                alpha=0.35,
            )

        if annotate:
            for _, r in sub.iterrows():
                label = f"G={int(r['G'])}, Y={int(r['Y'])}"
                if pair_col is not None:
                    label = f"{r[pair_col].replace('_to_', '→')}\n{label}"
                ax.annotate(
                    label,
                    (r["dist_mean"], r["metric_mean"]),
                    xytext=(4, 4),
                    textcoords="offset points",
                    fontsize=8,
                )

        ax.set_xlim(xmin, xmax)
        ax.set_title(f"{xmin:.2f} ≤ x ≤ {xmax:.2f}")
        ax.set_xlabel(xlabel)

    # hide unused axes
    for ax in axes[len(x_sections):]:
        ax.set_visible(False)

    # y label on left column
    for i, ax in enumerate(axes):
        if ax.get_visible() and (i % ncols == 0):
            ax.set_ylabel(ylabel)

    fig.suptitle(suptitle)

    # first tighten the subplot grid
    plt.tight_layout(rect=[0, 0, 0.88, 0.95])

    # then add a dedicated colorbar axis
    if scatter_ref is not None and color_by is not None:
        cax = fig.add_axes([0.90, 0.15, 0.02, 0.7])
        cbar = fig.colorbar(scatter_ref, cax=cax)
        cbar.set_label(color_by)

    return fig, axes, summary
```
