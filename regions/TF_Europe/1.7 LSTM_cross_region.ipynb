```python
# Build global scalers:
def build_global_scalers(
    res_xreg: dict,
    monthly_cols: list[str],
    static_cols: list[str],
    glacier_col: str = "GLACIER",
):
    Xm_all = []
    Xs_all = []

    df_train = res_xreg["df_train"]  # ('CH')
    df_test = res_xreg["df_test"]  # (all non-CH)

    # --- climate (row-level) ---
    Xm_all.append(df_train[monthly_cols].to_numpy(dtype=np.float64))
    Xm_all.append(df_test[monthly_cols].to_numpy(dtype=np.float64))

    # --- topo (glacier-level, deduplicated) ---
    Xs_all.append(
        df_train.groupby(glacier_col)[static_cols].first().to_numpy(
            dtype=np.float64))
    Xs_all.append(
        df_test.groupby(glacier_col)[static_cols].first().to_numpy(
            dtype=np.float64))

    Xm_all = np.vstack(Xm_all)
    Xs_all = np.vstack(Xs_all)

    scaler_m = StandardScaler().fit(Xm_all)
    scaler_s = StandardScaler().fit(Xs_all)

    return scaler_m, scaler_s
```

```python
def plot_region_shift_vs_performance(
    df_metrics: pd.DataFrame,
    all_shifts: dict,
    complement_key: str = "",
    performance_cols: list[str] | None = None,
    distance_cols: list[str] | None = None,
):
    """
    Scatter plots of region-level domain shift vs transfer performance.
    One point per region, annotated with region name.

    Creates a grid: rows = performance metrics, cols = distance metrics.
    Spearman rho is annotated on each subplot.

    Parameters
    ----------
    df_metrics : pd.DataFrame
        Output of evaluate_all_models. Index is the full key
        (e.g. "XREG_CH_TO_NOR"). Print df_metrics to check column names.
    all_shifts : dict
        Keys like "XREG_CH_TO_NOR", values from compute_domain_shift.
    complement_key : str
        Prefix prepended to df_metrics index to match all_shifts keys.
        Use "" (default) if df_metrics is already indexed by full keys
        like "XREG_CH_TO_NOR". Use "XREG_CH_TO_" if indexed by "NOR".
    performance_cols : list[str] or None
        Columns from df_metrics to use as y-axis. If None, auto-detects
        any column containing "rmse" or "bias" (case-insensitive).
    distance_cols : list[str] or None
        Shift dict keys for x-axis. Defaults to joint + climate + topo
        for both MMD2 and energy distance.

    Returns
    -------
    (fig, df_region) : tuple
        fig: matplotlib Figure
        df_region: flat DataFrame with one row per region, all metrics merged
    """
    from scipy import stats

    # --- auto-detect performance columns if not provided ---
    if performance_cols is None:
        performance_cols = [
            c for c in df_metrics.columns
            if any(kw in c.lower() for kw in ["rmse", "bias"])
        ]
        if not performance_cols:
            raise ValueError(
                f"No performance columns auto-detected in df_metrics.\n"
                f"Available columns: {list(df_metrics.columns)}\n"
                f"Pass performance_cols= explicitly.")
        print(f"Auto-detected performance columns: {performance_cols}")

    if distance_cols is None:
        distance_cols = [
            "D_mmd2_joint",
            "D_mmd2_climate",
            "D_mmd2_topo",
            "D_energy_joint",
        ]

    # --- build flat region DataFrame ---
    records = []
    for full_key in df_metrics.index:
        shift_key = f"{complement_key}{full_key}" if complement_key else full_key
        if shift_key not in all_shifts:
            print(f"Warning: '{shift_key}' not in all_shifts, skipping.")
            continue

        shift = all_shifts[shift_key]
        region_label = full_key.replace("XREG_CH_TO_", "")

        row = {"full_key": full_key, "region": region_label}
        for pc in performance_cols:
            row[pc] = df_metrics.loc[full_key, pc]
        for dc in distance_cols:
            row[dc] = shift.get(dc, float("nan"))
        records.append(row)

    if not records:
        raise ValueError(
            "No matching regions between df_metrics and all_shifts.\n"
            f"df_metrics index: {list(df_metrics.index)}\n"
            f"all_shifts keys:  {list(all_shifts.keys())}")

    df_region = pd.DataFrame(records)
    n_regions = len(df_region)
    print(f"Plotting {n_regions} regions: {list(df_region['region'])}")

    # --- colours ---
    colors = get_cmap_hex(cm.batlow, max(n_regions, 4))
    color_map = {r: colors[i] for i, r in enumerate(df_region["region"])}

    xlabel_map = {
        "D_mmd2_joint": "MMD2 joint",
        "D_mmd2_climate": "MMD2 climate",
        "D_mmd2_topo": "MMD2 topo",
        "D_energy_joint": "Energy distance joint",
        "D_energy_climate": "Energy distance climate",
        "D_energy_topo": "Energy distance topo",
    }

    nrows = len(performance_cols)
    ncols = len(distance_cols)

    fig, axes = plt.subplots(
        nrows,
        ncols,
        figsize=(4.5 * ncols, 4.2 * nrows),
        squeeze=False,
    )

    for r, pc in enumerate(performance_cols):
        for c, dc in enumerate(distance_cols):
            ax = axes[r][c]

            x = df_region[dc].values.astype(float)
            y = df_region[pc].values.astype(float)
            mask = np.isfinite(x) & np.isfinite(y)

            # --- one point per region ---
            for _, row in df_region.iterrows():
                xi = float(row[dc])
                yi = float(row[pc])
                if not (np.isfinite(xi) and np.isfinite(yi)):
                    continue
                ax.scatter(
                    xi,
                    yi,
                    color=color_map[row["region"]],
                    s=140,
                    zorder=3,
                    edgecolors="white",
                    linewidths=0.6,
                )
                ax.annotate(
                    row["region"],
                    (xi, yi),
                    xytext=(6, 4),
                    textcoords="offset points",
                    fontsize=9,
                    fontweight="bold",
                    color=color_map[row["region"]],
                )

            # --- Spearman correlation ---
            n_valid = mask.sum()
            if n_valid >= 3:
                rho, pval = stats.spearmanr(x[mask], y[mask])
                corr_txt = f"rho = {rho:.2f}  p = {pval:.2f}  n = {n_valid}"
            else:
                corr_txt = f"n = {n_valid} (too few for rho)"

            ax.text(
                0.04,
                0.97,
                corr_txt,
                transform=ax.transAxes,
                va="top",
                ha="left",
                fontsize=8,
                bbox=dict(boxstyle="round,pad=0.3",
                          fc="white",
                          ec="none",
                          alpha=0.8),
            )

            ax.set_xlabel(xlabel_map.get(dc, dc), fontsize=9)
            ax.set_ylabel(pc, fontsize=9)
            ax.grid(color="#e0e0e0", linewidth=0.5)
            ax.spines[["top", "right"]].set_visible(False)
            ax.set_axisbelow(True)

    fig.suptitle(
        "Region-level domain shift vs transfer performance\n"
        "(CH -> target regions, zero-shot)",
        fontsize=13,
        y=1.01,
    )
    plt.tight_layout()
    return fig, df_region


# --- Example usage ---
#
# # Step 1: always check column names and index first
# print(df_metrics)
#
# # Step 2: df_metrics index is already full keys like "XREG_CH_TO_NOR"
# # so complement_key="" (default)
# fig, df_region = plot_region_shift_vs_performance(
#     df_metrics=df_metrics,
#     all_shifts=all_shifts,
# )
# plt.show()
#
# # Step 3: narrow down once you know the column names
# fig, df_region = plot_region_shift_vs_performance(
#     df_metrics=df_metrics,
#     all_shifts=all_shifts,
#     performance_cols=["RMSE_a", "Bias_a"],          # <- actual column names
#     distance_cols=["D_mmd2_joint", "D_energy_joint"],
# )
# plt.show()
```

```python
def compute_glacier_level_shifts(
    res_all_xreg: dict,
    monthly_cols: list[str],
    static_cols: list[str],
    scaler_m,
    scaler_s,
    glacier_col: str = "GLACIER",
) -> pd.DataFrame:
    """
    For each XREG transfer key, compute per-glacier domain shift between
    the CH training set and each individual target glacier.

    For each glacier in the target region, MMD² and energy distance are
    computed between:
      - df_train (all CH rows)
      - all rows in df_test belonging to that glacier

    Parameters
    ----------
    res_all_xreg : dict
        Keys like "XREG_CH_TO_NOR", values with df_train / df_test.
    monthly_cols : list[str]
        Dynamic/climate feature columns (should include POINT_BALANCE if desired).
    static_cols : list[str]
        Static/topo feature columns.
    scaler_m, scaler_s : StandardScaler
        Pre-fitted scalers (fit on pooled train+test to ensure consistency).
    glacier_col : str
        Column identifying individual glaciers.

    Returns
    -------
    pd.DataFrame
        One row per (key, glacier) with columns:
        region, glacier, n_rows,
        D_mmd2_joint, D_mmd2_climate, D_mmd2_topo,
        D_energy_joint, D_energy_climate, D_energy_topo
    """
    rows = []

    for key, res in tqdm(res_all_xreg.items(), desc="Glacier-level shifts"):
        df_train = res.get("df_train")
        df_test  = res.get("df_test")

        if df_train is None or df_test is None or len(df_test) == 0:
            continue

        region = key.split("_TO_")[-1] if "_TO_" in key else key
        glaciers = df_test[glacier_col].dropna().unique()

        for glacier in glaciers:
            df_glacier = df_test[df_test[glacier_col] == glacier]
            n_rows = len(df_glacier)

            # skip glaciers with too few rows for a meaningful distance estimate
            if n_rows < 10:
                continue

            shift = compute_domain_shift(
                df_src=df_train,
                df_tgt=df_glacier,
                monthly_cols=monthly_cols,
                static_cols=static_cols,
                scaler_m=scaler_m,
                scaler_s=scaler_s,
                compute_marginals=False,  # keep fast at glacier level
            )

            rows.append({
                "key":              key,
                "region":           region,
                "glacier":          glacier,
                "n_rows":           n_rows,
                "D_mmd2_joint":     shift["D_mmd2_joint"],
                "D_mmd2_climate":   shift["D_mmd2_climate"],
                "D_mmd2_topo":      shift["D_mmd2_topo"],
                "D_energy_joint":   shift["D_energy_joint"],
                "D_energy_climate": shift["D_energy_climate"],
                "D_energy_topo":    shift["D_energy_topo"],
            })

    return pd.DataFrame(rows)


def compute_glacier_level_performance(
    preds_by_key: dict,
    glacier_col: str = "GLACIER",
    target_col: str = "target",   # actual column name in preds_by_key
    pred_col: str = "pred",
    min_rows_nrmse: int = 15,
) -> pd.DataFrame:
    """
    Compute per-glacier RMSE, NRMSE (normalised by observed PMB std) and bias.

    NRMSE = RMSE / std(observed PMB per glacier).
    A value of 1.0 means the model performs no better than predicting the
    glacier mean. NaN is returned for glaciers with fewer than min_rows_nrmse
    observations, where the std estimate is too noisy to be reliable.

    Parameters
    ----------
    preds_by_key : dict
        Keys like "XREG_CH_TO_NOR", values are DataFrames with
        at least glacier_col, target_col, pred_col columns.
    glacier_col : str
        Column identifying individual glaciers.
    target_col : str
        Column with observed PMB values.
    pred_col : str
        Column with predicted PMB values.
    min_rows_nrmse : int
        Minimum number of predictions required to compute a reliable NRMSE.
        Glaciers below this threshold get NaN for NRMSE (but still get RMSE).

    Returns
    -------
    pd.DataFrame
        One row per (key, glacier) with columns:
        region, glacier, n_preds, rmse, nrmse, bias
    """
    rows = []

    for key, df_preds in preds_by_key.items():
        if df_preds is None or len(df_preds) == 0:
            continue

        region = key.split("_TO_")[-1] if "_TO_" in key else key

        if glacier_col not in df_preds.columns:
            print(f"Warning: '{glacier_col}' not found in preds for {key}, skipping.")
            continue

        # --- validate target and pred columns, with helpful diagnostics ---
        if target_col not in df_preds.columns:
            print(
                f"Warning: target_col='{target_col}' not found in preds for {key}.\n"
                f"  Available columns: {list(df_preds.columns)}\n"
                f"  Pass the correct target_col= to compute_glacier_level_performance."
            )
            continue
        if pred_col not in df_preds.columns:
            print(
                f"Warning: pred_col='{pred_col}' not found in preds for {key}.\n"
                f"  Available columns: {list(df_preds.columns)}\n"
                f"  Pass the correct pred_col= to compute_glacier_level_performance."
            )
            continue

        for glacier, grp in df_preds.groupby(glacier_col):
            n = len(grp)
            residuals = grp[pred_col] - grp[target_col]
            rmse = float(np.sqrt((residuals ** 2).mean()))
            bias = float(residuals.mean())

            # NRMSE: only reliable with enough observations to estimate std
            pmb_std = float(grp[target_col].std())
            if n >= min_rows_nrmse and pmb_std > 0:
                nrmse = rmse / pmb_std
            else:
                nrmse = float("nan")

            rows.append({
                "key":      key,
                "region":   region,
                "glacier":  glacier,
                "n_preds":  n,
                "rmse":     rmse,
                "nrmse":    nrmse,
                "bias":     bias,
                "pmb_std":  pmb_std,  # keep for reference / diagnostics
            })

    return pd.DataFrame(rows)


def build_glacier_shift_vs_performance(
    res_all_xreg: dict,
    preds_by_key: dict,
    monthly_cols: list[str],
    static_cols: list[str],
    scaler_m,
    scaler_s,
    glacier_col: str = "GLACIER",
    target_col: str = "target",   # actual column name in preds_by_key
    pred_col: str = "pred",
    min_rows: int = 10,
    min_rows_nrmse: int = 20,
) -> pd.DataFrame:
    """
    Convenience wrapper: computes glacier-level shifts and performance,
    then merges them into a single DataFrame ready for plotting.

    Returns
    -------
    pd.DataFrame
        One row per (key, glacier) with both distance and performance columns.
        Rows with fewer than min_rows predictions are dropped.
    """
    df_shifts = compute_glacier_level_shifts(
        res_all_xreg=res_all_xreg,
        monthly_cols=monthly_cols,
        static_cols=static_cols,
        scaler_m=scaler_m,
        scaler_s=scaler_s,
        glacier_col=glacier_col,
    )

    df_perf = compute_glacier_level_performance(
        preds_by_key=preds_by_key,
        glacier_col=glacier_col,
        target_col=target_col,
        pred_col=pred_col,
        min_rows_nrmse=min_rows_nrmse,
    )

    if df_perf.empty:
        raise ValueError(
            "compute_glacier_level_performance returned an empty DataFrame. "
            "This usually means target_col or pred_col are wrong.\n"
            f"  target_col='{target_col}', pred_col='{pred_col}'\n"
            "  Check the warnings above and pass the correct column names."
        )

    df = pd.merge(
        df_shifts,
        df_perf,
        on=["key", "region", "glacier"],
        how="inner",
    )

    df = df[df["n_preds"] >= min_rows].copy()
    print(f"Glacier-level dataset: {len(df)} glaciers across {df['region'].nunique()} regions.")
    return df


def plot_glacier_shift_vs_performance(
    df: pd.DataFrame,
    distance_col: str = "D_mmd2_joint",
    performance_col: str = "rmse",
    size_col: str = "n_preds",
    hue_col: str = "region",
):
    """
    Scatter plot of glacier-level domain distance vs RMSE,
    with point size proportional to number of predictions and
    colour by region. Includes a Spearman correlation annotation.

    Parameters
    ----------
    df : pd.DataFrame
        Output of build_glacier_shift_vs_performance.
    distance_col : str
        Column to use as x-axis (domain shift metric).
    performance_col : str
        Column to use as y-axis (e.g. rmse or bias).
    size_col : str
        Column to scale marker size by.
    hue_col : str
        Column to colour points by.
    """
    from scipy import stats

    fig, ax = plt.subplots(figsize=(10, 6))

    regions = df[hue_col].unique()
    colors = get_cmap_hex(cm.batlow, len(regions))
    color_map = dict(zip(sorted(regions), colors))

    # size scaled by number of predictions (sqrt for visual balance)
    sizes = 20 + 80 * (df[size_col] - df[size_col].min()) / (
        df[size_col].max() - df[size_col].min() + 1e-9
    )

    for region, grp in df.groupby(hue_col):
        idx = grp.index
        ax.scatter(
            grp[distance_col],
            grp[performance_col],
            s=sizes[idx],
            color=color_map[region],
            alpha=0.7,
            edgecolors="none",
            label=region,
        )

        # optionally annotate glacier names for small groups
        if len(grp) <= 10:
            for _, row in grp.iterrows():
                ax.annotate(
                    row["glacier"],
                    (row[distance_col], row[performance_col]),
                    fontsize=6,
                    alpha=0.6,
                    xytext=(3, 3),
                    textcoords="offset points",
                )

    # Spearman correlation across all glaciers
    rho, pval = stats.spearmanr(df[distance_col], df[performance_col])
    ax.text(
        0.02, 0.97,
        f"Spearman ρ = {rho:.2f}  (p = {pval:.3f})  n = {len(df)}",
        transform=ax.transAxes,
        va="top", ha="left",
        fontsize=9,
        bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="none", alpha=0.8),
    )

    xlabel_map = {
        "D_mmd2_joint":   "MMD² joint",
        "D_mmd2_climate": "MMD² climate",
        "D_mmd2_topo":    "MMD² topo",
        "D_energy_joint": "Energy distance joint",
    }
    ylabel_map = {
        "rmse":  "RMSE [m w.e.]",
        "nrmse": "NRMSE [RMSE / σ(PMB)]",
        "bias":  "Bias [m w.e.]",
    }

    ax.set_xlabel(xlabel_map.get(distance_col, distance_col))
    ax.set_ylabel(ylabel_map.get(performance_col, performance_col))
    ax.set_title("Glacier-level domain shift vs transfer performance\n(CH → target regions, zero-shot)")
    ax.legend(title="Region", frameon=False, bbox_to_anchor=(1.01, 1), loc="upper left")
    ax.grid(color="#e0e0e0", linewidth=0.5)
    ax.spines[["top", "right"]].set_visible(False)

    plt.tight_layout()
    return fig
```
